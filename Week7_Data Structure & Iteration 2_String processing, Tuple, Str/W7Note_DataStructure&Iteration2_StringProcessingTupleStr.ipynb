        <ul><ul>
            <h4>A. 리스트가 지원하는 함수 (일부)</h4>
            <ul><ul>
                <li style="list-style-type:disc;"><strong>list.insert(i, x)</strong> : 주어진 i라는 인덱스 위치에 x라는 항목을 삽입하고 본래 i번째 이후의 인덱스 위치의 항목을 i+1번째 이후로 미루는 함수이다. 그래서 list.insert(0, x)는 리스트의 제일 첫 자리에 x를 삽입하는 함수이고, list.inset(len(list), x)는 list.append(x)와 같은 작업을 하는 함수이다.</li>
                <li style="list-style-type:disc;"><strong>list.remove(x)</strong> : 리스트에서 값이 x와 같은 항목을 인덱스 0번 부터 오름차순으로 검색하여 나오는 첫번째 x 항목을 삭제하고 그 다음 인덱스들을 한 칸씩 앞으로 이동시키는 함수이다. 그런 항목이 없으면 Value Error를 일으킨다.</li>
                <li style="list-style-type:disc;"><strong>list.clear( )</strong> : list의 모든 항목을 삭제해서 빈 리스트를 만드는 함수이다. del.list[:]와 같은 작업을 하는 함수이다.</li>
                <li style="list-style-type:disc;"><strong>list.sort(key=None, reverse=False)</strong> : 리스트의 항목들을 알파벳 오름차순 혹은 숫자 오름차순으로 정렬하는 함수이다. reverse를 True로 바꾸면 내림차순으로 정렬하게 된다. </li>
                <li style="list-style-type:disc;"><strong>list.reverse( )</strong> : 리스트의 요소들의 순서를 뒤집고 None를 return하는 함수이다. 다시말해 0번 리스트가 -1번 리스트로, 1번 리스트가 -2번 리스트로 …..해서 요소들의 순서를 역으로 바꿔주는 함수이다.</li>
                <li style="list-style-type:disc;"><strong>list.copy( )</strong> : 리스트의 얕은 사본(shallow copy, 껍데기만 복사한 것)을 돌려준다. list[:]와 같은 작업을 하는 함수이다. (얕은 사본, shallow copy의 반대말은 깊은 사본, deep copy로 내부까지 전부 복사한 것을 말한다.)</li>

        <h3>2. 스트링(str)</h3>
        <ul><ul>
            <h4>A. 문자열을 나타내는 방법을 스트링(str)이라고 한다.</h4>
            <ul><ul>
                <li style="list-style-type:disc;">‘abc’ 와 같이 써도 되고, “abc”와 같이 써도 된다. 의미상으로는 동일하지만 한 프로그램 내에서는 일관성 있게 사용하는 것을 권장한다.</li>
                <li style="list-style-type:disc;">여러줄에 걸친 문자열을 나타내고 싶다면, ‘’’abc def’’’나 “””abc def”””와 같이 따옴표를 3번 중첩하여 사용한다. 의미상으로는 ‘’’나 “””나 동일하지만, ‘’’는 주석에서 사용되는 경우가 더 많기 때문에, “””을 사용하는 것을 추천한다.</li>
                <li style="list-style-type:disc;">사실 여러줄에 걸친 문자열을 표현할 때는 ‘’’abc def’’’나 “””abc def”””보다는 ‘abc\ndef’가 더 선호되는 편이다. 이때, \n은 줄바꿈 예약어이다.</li>
            </ul></ul>

        <h3>3. 튜플(tuple)</h3>
        <ul><ul>
            <h4>A. 튜플(tuple)</h4>
            <ul><ul>
                <li style="list-style-type:disc;">정의: immutable sequences(수정 불가능한 시퀸스)이다. typically used to store collections of homogeneous items.</li>
                <ul><ul>
                    <li style="list-style-type:circle;">immutable: 원소의 추가/삭제, 값의 변경이 불가능하다.</li>
                    <li style="list-style-type:circle;">원소의 자료형: 임의의 자료형의 원소를 저장할 수 있다.</li>
                </ul></ul>
                <li style="list-style-type:disc;">인덱스(index): 음이 아닌 정수로 원소의 순서를 나타낸다.</li>
                <li style="list-style-type:disc;">표기법: ( )을 사용하고, 원소는 ,로 구분하여 정의한다.</li>
                <ul><ul>
                    <li style="list-style-type:circle;">정의할 때, ( )는 생략 가능하다. 또한, 원소가 하나일 때는 Expression(수식)과 구분하기 위해서 ,를 뒤에 추가하여야 한다.</li>
                </ul></ul>
                <li style="list-style-type:disc;">리스트와 유사한 점</li>
                <ul><ul>
                    <li style="list-style-type:circle;">인덱싱: 튜풀도 인덱싱을 할 때는 []을 이용한다.</li>
                    <li style="list-style-type:circle;">튜플은 리스트와 비슷한 연산 혹은 함수를 사용한다. 다만 원소의 변경은 없기 때문에, append, insert 등의 함수는 사용하지 못한다.</li>
                </ul></ul>
            </ul></ul>

        <h3>9. 가변 언패킹(unpacking)</h3>
        <ul><ul>
            <li style="list-style-type:disc;">튜플/리스트에 있는 값들을 각각 변수에 대입하는 것을 언패킹이라 한다.</li>
            <li style="list-style-type:disc;">앞에서도 말했듯, 언패킹 때는 반드시 원소의 개수와 변수의 개수가 일치하여야 한다.</li>
                <ul><ul>

        <h3>3. 제어 흐름: break, continue, pass(반복문의 else절의 필요성을 느끼게 해주는 제어 흐름 조절 예약어)</h3>
        <ul><ul>
            <h4>A. break: 가장 안쪽의 for 혹은 while문의 실행을 끝내는 명령</h4>
            <ul><ul>
                <li style="list-style-type:disc;">break문을 이용해 반복문의 실행을 끝내면 else절이 실행되지 않는다. (else절도 for 혹은 while문의 일부이기 때문)</li>
                <li style="list-style-type:disc;">예제</li>

                <li style="list-style-type:disc;">for에서 continue를 만날 경우: continue 다음의 코드블럭을 실행하지 않고, sequence의 다음 값을 loop variable에 대입하고 for문 내부의 코드블럭을 다시 실행한다.</li>
