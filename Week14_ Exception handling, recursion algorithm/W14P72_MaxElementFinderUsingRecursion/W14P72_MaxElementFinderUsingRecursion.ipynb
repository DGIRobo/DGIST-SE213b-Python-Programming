### <span style='font-family:Georgia'> 함수 find_max(t, l, r)
<br>

> ##### <li> <span style='font-family:Georgia'> 함수동작 </li>
```Bash
t라는 리스트에서 index l과 r사이에서의 최댓값을 반환함
```

> ##### <li> <span style='font-family:Georgia'> 매개변수 </li>
```Bash
t: 1개 이상의 원소를 포함하고 있는 리스트 (이 조건에 맞지 않는 경우는 신경쓰지 않아도 됨)
l: (int) 최댓값을 찾기 위한 범위의 최소 인덱스 (소문자 L)
r: (int) 최댓값을 찾기 위한 범위의 최대 인덱스
```
    
> ##### <li> <span style='font-family:Georgia'> 반환값 </li>
```Bash
t[l], t[l+1], t[l+2], ..., t[r]에서 최대값을 찾아서 반환
```

### <span style='font-family:Georgia'> 참고

t라는 리스트가 있을 때, 아래와 같이 find_max() 함수를 호출하면 전체 범위에서 최대값을 찾아 반환하게 된다. (인덱스는 0부터 len(t)-1까지임)

find_max(t, 0, len(t)-1)

### <span style='font-family:Georgia'> Function Usage Example

````python
# Answer of Problem 
def find_max(t, l, r):
    """Retuan the max number in t[l:r-1]"""
    #Base case
    if l == r:
        return t[l]
    #Recursive case
    else:
        candidate = find_max(t, l+1, r)
        if candidate < t[l]:
            return t[l]
        else:
            return candidate

if __name__ == '__main__':
    t = [1, 2]
    print(find_max(t, 0, len(t) - 1))
    t = [36, 43, 70, 76, 32, 7, 8, 21, 69, 78]
    print(find_max(t, 0, len(t) - 1))
````
