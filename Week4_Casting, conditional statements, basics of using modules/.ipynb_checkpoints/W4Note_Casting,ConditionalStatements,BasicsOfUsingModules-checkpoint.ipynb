        <ul><ul>
            <p>큰 문제를 해결 가능한 작은 문제로 나누는 것, 경우에 따라 여러 번 반복.</p>
        </ul></ul>
        <h3>2. 패턴인식 (Pattern recognition)</h3>
        <ul><ul>
            <p>(작게 나뉘어진) 주어진 문제에서 공통된 요소(패턴)을 찾음.</p>
        </ul></ul>
        <h3>3. 추상화 (Abstraction)</h3>
        <ul><ul>
            <p>문제 해결에 있어 중요한 공통된 부분만을 놔두고, 다른 구체적인 사항들을 없앰.</p>
        </ul></ul>
        <h3>4. 알고리즘 (Algorithms)</h3>
        <ul><ul>
            <p>어떤 입력에도 원하는 결과를 얻을 수 있도록 잘 정의된 연산 단계를 만든다. ex) 덧셈, 뺄셈, 곱셈, 나눗셈</p>
        </ul></ul>

    <h2>ⅱ. 형변환(type conversion)</h2>
    <ul><ul>
        <h3>1. 형변환 (type conversion)</h3>
        <ul><ul>
            <li style="list-style-type:disc;">한 자료형(data type)에서 다른 자료형으로 변경하는 것을 형변환(type conversion)이라 부른다.</li>
            <li style="list-style-type:disc;">input()으로 입력 받은 값은 문자열로 반환되기 때문에 받은 값을 사용하여 원하는 연산(예: 사칙연산)을 수행하기 위해서는 형변환이 필요하다.</li>
        </ul></ul>
        <h3>2. 형변환 함수: 바뀔 자료형의 이름을 함수처럼 사용한다.</h3>
        <ul><ul>
            <h4>A. int( )</h4>
            <ul><ul>
                <li style="list-style-type:disc;">( ) 안의 인자를 정수형으로 형변환 시키는 함수이다.</li>
                <li style="list-style-type:disc;">예시</li>
                <ul><ul>
                    <li style="list-style-type:circle;">integer=int(3.14)로 입력하면, integer에 3이 저장된다.</li>
                    <li style="list-style-type:circle;">integer=int(‘abc’)로 입력하면, 오류가 발생한다.</li>
                    <li style="list-style-type:circle;">integer=int(‘3.7’)로 입력하면, integer에 3이 저장된다.</li>
                </ul></ul>
            </ul></ul>
            <h4>B. float( )</h4>
            <ul><ul>
                <li style="list-style-type:disc;">( ) 안의 인자를 실수형으로 형변환 시키는 함수이다.</li>
                <li style="list-style-type:disc;">예시</li>
                <ul><ul>
                    <li style="list-style-type:circle;">real=float(3)로 입력하면, real에 3.0이 저장된다.</li>
                    <li style="list-style-type:circle;">real=float(‘abc’)로 입력하면, 오류가 발생한다.</li>
                    <li style="list-style-type:circle;">real=float(‘3.7’)로 입력하면, real에 3.7이 저장된다.</li>
                </ul></ul>
            </ul></ul>
            <h4>C. 형변환이 가능하지 않은 경우에는 오류가 발생한다.</h4>
            <ul><ul>
                <p>예를 들면, 문자열 ‘abc’는 정수 혹은 실수로 변환되지 않는다.</p>
            </ul></ul>
        </ul></ul>
        <h3>3. 자료형 확인: type( )</h3>

    <h2>ⅲ. 제어 흐름 (control flow)</h2>
    <ul><ul>
        <h3>1. 명령문 (statement)</h3>
        <ul><ul>
            <li style="list-style-type:disc;">실행의 단위이다.</li>
            <li style="list-style-type:disc;">보통 한 줄의 코드로 나타내어진다.</li>
            <li style="list-style-type:disc;">명령문은 한 줄씩 차례로 실행된다.</li>
        </ul></ul>
        <h3>2. 제어 흐름 (control flow)은 명령문들이 실행되는 순서/방법을 지시한다.</h3>
        <ul><ul>
            <li style="list-style-type:disc;">조건문: 조건에 따라 다른 명령문들을 실행한다.</li>
            <li style="list-style-type:disc;">반복문: 하나 혹은 그 이상의 명령문들을 반복해서 실행한다.</li>
            <li style="list-style-type:disc;">함수: 동일한 명령문들을 프로그램의 여러 곳에서 실행한다..</li>
        </ul></ul>
    </ul></ul>

    <h2>ⅳ. 조건문: if, else, elif</h2>
    <ul><ul>
        <h3>1. 조건문과 불리언(boolen)식</h3>
        <ul><ul>
            <ul><ul>
                <li style="list-style-type:disc;">조건에 따라 다른 명령문을 실행해야 되는 경우가 있다.</li>
                <li style="list-style-type:disc;">많은 프로그래밍 언어에서 조건문은 불리언(boolen)식으로 표현해야한다.</li>
                <ul><ul>
                    <li style="list-style-type:circle;">불리언(boolen)식: True 혹은 False를 계산하는 식(Expression)</li>
                </ul></ul>
                <li style="list-style-type:disc;">다음과 같은 값들은 False로 인식된다.</li>
                <ul><ul>
                    <li style="list-style-type:circle;">false로 정의된 상수: None, False</li>
                    <li style="list-style-type:circle;">수를 나타내는 자료형에 0이 저장된 경우: 0, 0.0, 0j, Decimal(0), Fraction(0, 1)</li>
                    <li style="list-style-type:circle;">원소를 저장하고 있지 않은 자료구조: “”(내용이 없는 문자열), ( )(내용이 없는 튜플), [](내용이 없는 리스트), set( ), range(0)</li>
                </ul></ul>
                <li style="list-style-type:disc;">True of False로 인식될 수 있는 값에 무엇이 있는지는 <a href="https://docs.python.org/3/library/stdtypes.html?highlight=boolean#truth-value-testing">다음 사이트</a>를 통해 확인해볼 수 있다.</li>
            </ul></ul>
        </ul></ul>
        <h3>2. 불리언(boolen)식(expression) 조건문의 표현</h3>
        <ul><ul>
            <li style="list-style-type:disc;">비교 연산자와 불리언 연산자를 통해 조건문을 표현하면, 불리언식이 된다.</li>
            <li style="list-style-type:disc;">비교 연산자의 종류</li>
            <ul><ul>
                <table align="left" style="width:500px;margin-right:100%;">
                    <thead>
                        <tr><th>연산자</th><th>설명</th><th>예시</th><th>실행결과</th></tr>
                    </thead>
                    <tbody>
                        <tr><td>&#60;</td><td>Less than</td><td>5&#60;3<br>3&#60;5</td><td>False<br>True</td></tr>
                        <tr><td>&#62;</td><td>Greater than</td><td>5&#62;3<br>3&#62;5</td><td>True<br>False</td></tr>
                        <tr><td>&#60;=</td><td>Less than or equal to</td><td>3&#60;=6</td><td>True</td></tr>
                        <tr><td>&#62;=</td><td>Greater than or equal to</td><td>4&#62;=3</td><td>True</td></tr>
                        <tr><td>==</td><td>Equal to</td><td>2==2<br>‘str’==’stR’<br>‘str’==’str’</td><td>True<br>False<br>True</td></tr>
                        <tr><td>!=</td><td>Not equal to</td><td>‘str’!=’stR’</td><td>True</td></tr>
                    </tbody>
                </table>
            </ul></ul>
            <li style="list-style-type:disc;">불리언(Boolean) 연산자의 종류</li>
            <ul><ul>
                <table align="left" style="width:500px;margin-right:100%;">
                    <thead>
                        <tr><th>x</th><th>y</th><th>x and y</th><th>x or y</th><th>not x</th></tr>
                    </thead>
                    <tbody>
                        <tr><td>False</td><td>False</td><td>False</td><td>False</td><td>True</td></tr>
                        <tr><td>False</td><td>True</td><td>False</td><td>True</td><td></td></tr>
                        <tr><td>True</td><td>False</td><td>False</td><td>True</td><td>False</td></tr>
                        <tr><td>True</td><td>True</td><td>True</td><td>True</td><td></td></tr>
                    </tbody>
                </table>
                <ul><ul>
                    <li style="list-style-type:circle;"><strong>and 연산</strong>: 두 인자가 모두 True여야 True를 반환하고, 그 이외에는 False를 반환하기 때문에, 첫 번째 인자가 True인 경우에만, 두 번째 인자를 계산한다.</li>
                    <li style="list-style-type:circle;"><strong>or 연산</strong>: 두 인자가 모두 False여야 False를 반환하고, 그 이외에는 True를 반환하기 때문에, 첫 번째 인자가 False인 경우에만, 두 번째 인자를 계산한다. </li>
                    <li style="list-style-type:circle;"><strong>not 연산</strong>: not 연산자는 불리언(boolen) 연산자 이외의 연산자보다 우선 순위가 낮다. 따라서 not a==b는 not (a==b)의 순서로 계산된다. </li>
                    <li style="list-style-type:circle;">또한, 이 때문에 a==not b로 적으면 문법오류가 발생한다. (a==not) b로 읽혀지기 때문이다. a==(not b)를 연산할 의도였으면, a==(not b)로 적어주어야 오류가 발생하지 않는다.</li>
                </ul></ul>
            </ul></ul>

    <h2>ⅴ. 코드블럭(Code block)</h2>
    <ul><ul>
        <h3>1. 코드블럭이란?</h3>
        <ul><ul>
            <p><strong>하나의 단위로 실행되는</strong> <strong>하나 이상의 명령문들</strong>을 말한다.</p>
        </ul></ul>
        <h3>2. python에서는 빈 칸으로 코드블럭을 표시한다.</h3>
        <ul><ul>
            <li style="list-style-type:disc;">탭 문자 혹은 빈 칸으로 들여쓰기를 만들어 코드블럭을 표시하는데, 공통된 코드블럭 내에서는 같은 정도의 들여쓰기를 사용해야 한다.</li>
            <li style="list-style-type:disc;">만약 statement1은 4개 빈칸으로 들여쓰기하고, statement2는 2개 빈칸으로 들여쓰기하면, 컴퓨터는 statement1까지만 코드블럭으로서 인식한다.</li>
            <li style="list-style-type:disc;">들여쓰기는 4개의 빈 칸을 사용하는 것을 강력하게 권장한다. 빈 줄이 있을 경우 그 줄은 무시된다.</li>
        </ul></ul>

    <h2>ⅵ. 모듈 사용: math, random</h2>
    <ul><ul>
        <h3>1. Python에서 모듈/패키지 소개</h3>
        <ul><ul>
            <li style="list-style-type:disc;">Python에는 수천~수만 종류의 패키지가 존재한다. 따라서 이미 만들어진 모듈/패키지를 사용하면 간단하게 프로그램을 작성할 수 있다.</li>
            <li style="list-style-type:disc;">모듈 (module): 재사용하고자 하는 변수나 함수의 정의문들을 파일로 저장한 것으로, 다른 프로그램이나 쉘 환경에서 호출하여 사용할 수 있다.</li>
            <li style="list-style-type:disc;">패키지 (package): 보통 계층구조를 가지는 여러 모듈들을 모아 놓은 것을 패키지라고 한다.</li>
            <ul><ul>
                <li style="list-style-type:circle;"><a href="https://pypi.org/">전체 리스트</a></li>
                <li style="list-style-type:circle;"><a href="https://docs.pytho.org/3/library/">Python Standard Library</a></li>
                <li style="list-style-type:circle;"><a href="https://wiki.python.org/moin/UsefulModules">유용한 패키지 리스트</a></li>
            </ul></ul>
        </ul></ul>
        <h3>2. 모듈을 사용하기 위해 알아야 하는 개념: 네임스페이스(namespace)</h3>
        <ul><ul>
            <li style="list-style-type:disc;">네임스페이스(namespace): ‘명칭공간’으로 번역되기도 하며, 변수, 함수 등이 정의되어 있는 공간을 지칭하는 표현이다.</li>
            <ul><ul>
                <li style="list-style-type:circle;">프로그램이 실행될 때, 컴퓨터는 전역 네임스페이스(global namespace)를 만들어 프로그램이 사용할 수 있도록 할당해준다.</li>
                <li style="list-style-type:circle;">프로그램에서 새로운 변수가 정의되면, 컴퓨터는 네임스페이스(namespace)에 변수의 값을 저장할 공간을 만든다.</li>
                <li style="list-style-type:circle;">프로그램의 실행 중 함수, 모듈을 불러오게 되면, 컴퓨터는 전역 네임스페이스(global namespace)와는 별개로 그 함수, 모듈만을 위한 독립적인 네임스페이스(이 네임스페이스의 이름에는 그 함수 혹은 모듈의 이름이 붙여진다.)를 만들어 그 함수, 모듈이 사용할 수 있도록 할당해준다.</li>
            </ul></ul>
        </ul></ul>
        <h3>3. 모듈/패키지 사용하기.</h3>
        <ul><ul>
            <h4>A. 모듈을 읽어들이는 방법</h4>
            <ul><ul>
                <li style="list-style-type:disc;">import module_name의 명령어를 통해 프로그램에 모듈을 읽어 들일 수 있다. </li>
                <li style="list-style-type:disc;">보통 이미 만들어진 모듈(module)들은 .py의 확장자명을 가지고 있을 텐데, 이때 불러들일 모듈(module)의 파일명 ~~.py에서 확장자명 .py를 뺀 ~~가 module_name이 된다.</li>
            </ul></ul>
            <h4>B. 모듈 사용방법</h4>
            <ul><ul>
                <li style="list-style-type:disc;">xxx가 변수이름인 경우, module_name.xxx와 같이 명령어를 적으면, 모듈에 속한 변수 중 이름이 xxx인 변수를 가져와 global namespace에 module_name.xxx로 저장한다.</li>
                <li style="list-style-type:disc;">xxx가 함수이름인 경우, module_name.xxx와 같이 명령어를 적으면, .모듈에 속한 함수 중 이름이 xxx인 함수를 가져와 실행한다.</li>
                <li style="list-style-type:disc;">사용예시</li>
                <ul><ul>
                    <li style="list-style-type:circle;">module_name.function_name( )</li>
                    <li style="list-style-type:circle;">module_name.variable_name</li>
                </ul></ul>
