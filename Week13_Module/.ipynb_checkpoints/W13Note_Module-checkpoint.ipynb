        <h3>1. 모듈의 사용법</h3>
        <ul><ul>
            <h4>A. 모듈/패키지 사용하기</h4>
            <ul><ul>
                <li style="list-style-type:disc;">모듈을 읽어들이는 방법</li>
                <ul><ul>
                    <li style="list-style-type:circle;">일반적으로 import module_name을 파일 맨 앞에 써서 파일이 해당 module_name의 함수 및 변수를 사용한다는 것을 알린다. 이렇게 쓰면, 컴퓨터는 global namespace에 module_name이라는 변수를 만들고, 이 변수에 해당파일이 저장된 공간(module_name이라는 namespace이다.)으로 가는 주소를 할당한다.</li>
                    <li style="list-style-type:circle;">여기서 module_name은 확장자 .py를 뺀 파일이름을 말한다.</li>
                </ul></ul>
                <li style="list-style-type:disc;">모듈(모듈 내의 변수 및 함수) 사용방법: module_name.xxx와 같이 사용하면 모듈에 속한 함수 혹은 변수 이름을 뜻한다. 사용예시는 다음과 같다.</li>
                <ul><ul>
                    <li style="list-style-type:circle;">module_name.function_name()</li>
                    <li style="list-style-type:circle;">module_name.variable_name</li>
                </ul></ul>
            </ul></ul>

            <h4>C. import 하는 여러가지 방법 1(모듈에서 원하는 값들만 읽어들이는 방법)</h4>
            <ul><ul>
                <li style="list-style-type:disc;">일반적인 방법 (모듈의 namespace를 만들어 거기에 모듈의 내용을 전부 import하는 방법)</li>
                <ul><ul>
                    <text>import math<br></text>
                    <text>print(math.pi, math.e)</text>
                </ul></ul>
                <li style="list-style-type:disc;">특정 객체(변수, 함수, 클래스 등)를 그 이름 그대로 import하는 방법</li>
                <ul><ul>
                    <text>from math import pi, e<br></text>
                    <text>print(pi, e)</text>
                </ul></ul>
                <li style="list-style-type:disc;">모든 객체(*표시가 모든을 의미한다.)를 그 이름 그대로 import하는 방법 (사용자가 정의한 변수, 함수, 클래스와 혼동이 일어나기 쉽기 때문에 권장하지 않는 방법이다.)</li>
                <ul><ul>
                    <text>from math import*<br></text>
                    <text>print(pi, e, tau)</text>
                </ul></ul>
            </ul></ul>
            <h4>D. import 하는 여러가지 방법 2(모듈의 이름을 줄여서 읽어들이는 방법)</h4>
            <ul><ul>
                <li style="list-style-type:disc;">모듈 이름을 간략하게 줄여서 가져오는 것이 가능함.: 단, 모듈 이름을 줄일때는 관례적인 경우를 따르는 것이 좋음.</li>
                <ul><ul>
                    <li style="list-style-type:circle;">예시1: import math as m<br>&#160;&#160;&#160;&#160;&#160;&#160;&#160;&#160;&#160;&#160;&#160;print(m.pi, m.e)</li>
                    <li style="list-style-type:circle;">예시2: import numpy as np</li>
                    <li style="list-style-type:circle;">예시3: import pandas as pd</li>
                </ul></ul>
                <li style="list-style-type:disc;">그런데 이렇게 모듈의 이름을 줄여서 가져오면, 그 모듈을 사용할 때는 줄인 이름만 사용할 수 있다. 즉, math를 m이라는 줄인이름으로 import했다면, m.pi와 같은 방식으로만 사용할 수 있다는 것이다. math.pi라고 쓰면 Error가 난다. 그 이유는 import math as m이라 쓰면, 컴퓨터는 global namespace에 m이라는 변수를 만들고 이 변수에 math라는 모듈이 저장된 namespace로의 주소(pointer, reference)를 할당하기 때문이다.</li>
            </ul></ul>
        </ul></ul>

        <h3>2. 모듈을 만드는 방법과 import의 의미</h3>
        <ul><ul>
            <li style="list-style-type:disc;">어떤 파이썬 파일이든 모듈로 사용하는 것이 가능하다.</li>
            <ul><ul>
                <li style="list-style-type:circle;">.py를 제외한 파일 이름이 모듈 이름이 된다.</li>
                <li style="list-style-type:circle;">파이썬 파일 안에서 정의한 변수, 함수, 클래스를 import해서 사용하는 것이 가능하다.</li>
            </ul></ul>
            <li style="list-style-type:disc;">모듈을 import하면 다음의 작업이 순서대로 시행된다.</li>
            <ul><ul>
                <li style="list-style-type:circle;">모듈을 위한 별도의 namespace가 생성된다.</li>
                <li style="list-style-type:circle;">모듈을 위한 namespace에서 import 대상인 파이썬 프로그램을 한 줄씩 읽으면서 실행한다.</li>
                <ul><ul>
                    <p>▶ 이 과정에서 변수, 함수, 클래스 등이 있으면, namespace 안에 해당 내용이 생성된다.</p>
                    <p>▶ 이 과정에서 화면 출력 등과 같은 입출력 명령을 만나면 그 명령을 그대로 실행한다.</p>
                </ul></ul>
            </ul></ul>
        </ul></ul>

        <h3>4. 패키지(Package)란?</h3>
        <ul><ul>
            <li style="list-style-type:disc;">하나의 Object를 만들기 위해서 관련된 Classes, Variables, Functions를 모두 모아놓은 파일이 module이었다.</li>
            <li style="list-style-type:disc;">그런데 이런 module을 여러 개를 한꺼번에 사용해야 되는 경우가 있다. 예를 들면 수학 연산과 기하학과 벡터연산을 한꺼번에 하고 싶게 되는 경우 math 모듈, 기하학 모듈, 벡터 모듈을 각각 import해서 가져와야 하는 것이다.</li>
            <li style="list-style-type:disc;">그런데 import해야 하는 파일이 상당히 많아지면 번거로워지므로, 관련된 module들을 하나로 묶어 패키지를 만들고 패키지를 가져오는 것으로 그 module들의 내용을 모두 쓸 수 있게 하였다. 이것이 패키지이다.</li>
            <img src="./Imgs for W13Note/package.png" alt="" loading="lazy"  align="left" style="width:250px;margin-right:100%;">
        </ul></ul>

        <h3>5. 패키지를 만드는 방법</h3>
        <ul><ul>
            <li style="list-style-type:disc;">패키지를 사용하고자 하는 python 파일이 있는 디렉토리에 패키지 파일을 만들어야 해당 python 파일에서 그 패키지를 사용할 수 있다.</li>
            <li style="list-style-type:disc;">패키지 파일 안에 사용하고자 하는 모듈 파일들을 넣으면, python 파일에서 그 패키지 파일을 import하는 것으로써 그 모듈파일들을 사용할 수 있게 된다.</li>
            <li style="list-style-type:disc;">예전에는 패키지 파일 안에 사용하고자 하는 module 파일들을 다같이 집어넣고 __init__.py라는 파일을 하나 더 만들어주어야 패키지로서 사용할 수 있었다. 그런데, 요즈음에는 __init__.py을 안만들어도 패키지 파일로 사용할 수 있다.</li>
            <li style="list-style-type:disc;">패키지 파일을 python 파일에서 사용하는 방법은 newpkg라는 패키지 안의 module1을 쓰고 싶다면, import newpkg.module1이라 쓰는 것이다.</li>
            <li style="list-style-type:disc;">newpkg 패키지 파일 안의 newminipkg 패키지 파일 안의 module1을 쓰고 싶다면, import newpkg.newminipkg.module1처럼 쓰면 된다. 즉, 패키지를 안에 패키지를 넣어서 만들어서 python 파일에서 접근하는것도 가능하다.</li>
        </ul></ul>

        <h3>6. 파이썬이 실행될 때 자동으로 만들어지는 전역변수: __name__</h3>
        <ul><ul>
            <li style="list-style-type:disc;">__name__: 파이썬 프로그램이 실행될 때, 어떤 파일을 가져와서 실행하고 있는지를 나타내는 전역변수</li>
            <ul><ul>
                <li style="list-style-type:circle;">개별적인 python 파일을 실행할 때(예: elice, pycharm, shell 등에서 실행): ‘__main__’이라는 값이 __name__에 할당됨.</li>
                <li style="list-style-type:circle;">한 파일이 모듈로 import되어 실행될 때: 모듈 함수가 실행될 때 ‘__moduleName__’이 __name__에 할당된다. 그 외에 개별적인 python 파일의 변수나 함수가 실행될때는 ‘__main__’이라는 값이 __name__에 할당된 상태가 된다.</li>
            </ul></ul>
            <li style="list-style-type:disc;">참고: 아래와 같이 모듈 함수 내에 if문을 작성해서 module을 import하지 않았을 때만 수행되는 테스트 코드 등을 넣는 것이 일반적임</li>
            <ul><ul>

            </ul></ul>
            <li style="list-style-type:disc;">예제: fibonacci.py 파일을 모듈로만 사용하는 것이 가능하도록 짠 코드</li>
            <ul><ul>

                <ul><ul>
                    <li style="list-style-type:circle;">module로 사용되어 fibonacci.fibonacci.list(n)으로 사용되면 if문은 함수 정의문 바깥에 있으므로 실행되지 않는다. 정의문 안에 위치하더라도 모듈로 사용된거여서 __name__은 ‘__fibonacci__’가 되어 if문은 false가 되어 실행되지 않는다.</li>
                    <li style="list-style-type:circle;">하지만, fibonacci.py를 직접 켜서 파일을 실행하면 if문까지 실행되어서 assert(fibonacci_list(10) == [1, 1, 2, 3, 5, 8])가 실행된다.</li>
                    <li style="list-style-type:circle;">이때, assert()는 ()안의 값이 True일 때, error를 발생하는 함수인데, fibonacci.py를 직접 켜서 파일을 실행하면 fibonacci_list(10)는 항상 [1, 1, 2, 3, 5, 8])이 되므로 if문이 무조건 실행되어 코드가 Error가 난다.</li>
                    <li style="list-style-type:circle;">이렇게 사용하는 것이 좋은 이유는 내가 module을 직접 만들 때, module내의 함수들이 제대로 동작하는지 테스트해야 하는데, 테스트 한 결과가 이 module로 사용할 때도 실행되면 안되므로, 이렇게 쓰는 것이다. 이렇게 쓰면 모듈로 사용할 때는 테스트코드가 실행되지 않기 때문이다.</li>

            <h4>B. random – Generate pseudo-random numbers</h4>
            <ul><ul>
                <li style="list-style-type:disc;">random module이란?: 숫자를 또는 문자 등을 랜덤으로 출력하는 함수들을 제공하는 모듈이다.</li>
                <li style="list-style-type:disc;">사용예시</li>
                <img src="./Imgs for W13Note/random module.png" alt="" loading="lazy"  align="left" style="width:500px;margin-right:100%;">
                <ul><ul>
                    <li style="list-style-type:circle;">random.random( ) 함수: 0보다 같거나 크고 1보다는 작은 float 값을 랜덤으로 반환하는 함수이다.</li>
                    <li style="list-style-type:circle;">random.uniform(a, b) 함수: a보다 같거나 크고 b보다는 작은 float 값을 랜덤으로 반환하는 함수이다.</li>
                    <li style="list-style-type:circle;">random.randrange(b) 함수: 0보다 같거나 크고 b보다는 작은 int 값을 랜덤으로 반환하는 함수이다.</li>
                    <li style="list-style-type:circle;">random.randrange(a, b) 함수: a보다 같거나 크고 b보다는 작은 int 값을 랜덤으로 반환하는 함수이다.</li>
                    <li style="list-style-type:circle;">random.randint(a, b) 함수: a보다 같거나 크고 b보다는 작거나 같은 int 값을 랜덤으로 반환하는 함수이다.</li>
                    <li style="list-style-type:circle;">random.choice(‘string’) 함수: string을 구성하는 문자 중에서 아무 문자나 랜덤으로 1개 뽑아서 반환하는 함수이다.</li>
                </ul></ul>
                <li style="list-style-type:disc;">이 ramdom 모듈의 함수들은 마치 random으로 값을 내는것 같지만 실제로 컴퓨터는 랜덤으로 값을 만드는 것이 불가능하다. 그래서 컴퓨터는 어떤 Operation을 통해 만든 값을 추출하는데 그 값이 마치 random인 것처럼 보이는 것이다. 대표적인 예시로 C언어는 random 함수가 시간값을 읽고 여기에 곱이나 차 등의 연산을 써서 반환값을 만드는 방식으로 동작해서 마치 랜덤값이 나오는 것처럼 보인다.</li>
                <li style="list-style-type:disc;">그런데, 이렇게 랜덤으로 출력되는 코드를 짜면 이 random값의 변화에 따라 코드가 오류가 날 수도 있고 정상동작할 수도 있다. 그래서 random 코드를 사용하면 코드의 유지, 보수, 디버깅이 힘든데, 그래서 이 모듈은 random.seed(a) 함수를 제공한다.</li>
                <li style="list-style-type:disc;">함수: 컴퓨터가 랜덤변수를 출력하는 연산의 기준은 a라는 숫자에 맞춰주는 것이다. 컴퓨터의 랜덤연산이 실은 규칙적인 계산이기 때문에 이렇게 맞춰주면 다음에 출력되는 숫자의 규칙이 직접적으로 보여서 다음에 출력되는 값을 유추할 수 있다. 그래서 랜덤변수를 사용한 코드는 이 함수를 사용해서 유지, 보수, 디버깅을 한다. 다음에 나올 값이 무엇인지 대강 알 수 있으므로, 그 값에 따라 오류를 찾기가 쉽기 때문이다.</li>

            <h4>B. 정규식(regular exprssion) 처리란?</h4>
            <ul><ul>
                <li style="list-style-type:disc;">어떤 웹의 텍스트를 긁어오거나 텍스트 파일의 텍스트를 긁어오거나 바이너리 파일의 텍스트를 긁어와서 데이터를 추출해보면, 추출이 어렵다는 것을 알 수 있다.</li>
                <li style="list-style-type:disc;">그 이유는 Contact: 가 쓰인 문장에서 Contact: 뒤에 쓰인 문장을 가져오라고 했는데, 어떤 파일에서는 Contact: 라 쓰여 있을지는 모르지만, _Contact: 또는 Contact :라고 쓰여있을 수도 있기 때문이다.</li>
                <li style="list-style-type:disc;">위 처럼 Contact: 가 쓰인 문장에서 Contact: 뒤에 쓰인 문장을 가져오라고 쓰면, _Contact: 또는 Contact :라고 쓰여있는 곳 뒤에 있는 전화번호는 들고오지 못하게 되는 것이다.</li>
                <li style="list-style-type:disc;">이제 이것을 해결하기 위해 쓰는 것이 정규식 처리 모듈이다.</li>
                <li style="list-style-type:disc;">정규식 처리 모듈은 문자의 패턴을 가지고 패턴에 맞는 문자열을 찾아내는 것을 도와준다. 예를 들면 전화번호는 숫자숫자숫자-숫자숫자숫자-숫자숫자숫자숫자 꼴의 패턴을 가진 문자열이므로 정규식 처리모듈을 써서 이 패턴을 정의하고 텍스트를 읽을 때 이 패턴에 맞는 문자열은 출력하라고 하면 앞에서 말한 문제가 발생하지 않고 전화번호들을 제대로 가져올 수 있게 되는 것이다.</li>
                <li style="list-style-type:disc;">한마디로 정규식 처리 모듈은 문자열 패턴을 정의하는 것을 도와주는 모듈이다.</li>
            </ul></ul>
            <h4>C. 정규식(regular expression) 처리의 방법 (문자열 패턴의 정의 방법)</h4>
            <ul><ul>
                <img src="./Imgs for W13Note/regular expression method.png" alt="" loading="lazy"  align="left" style="width:250px;margin-right:100%;">
                <ul><ul>
                    <li style="list-style-type:disc;">정규식 처리는 일단 re 모듈에서 지원하므로 re모듈을 import하고 시작한다.</li>
                    <li style="list-style-type:disc;">.은 문자열 패턴에서 한 글자를 의미한다.</li>
                    <li style="list-style-type:disc;">[012]라고 쓰면 그 자리에 있는 한 글자가 0또는 1또는 2로 쓰여진 패턴을 가지고 있다는 뜻이다.</li>
                    <li style="list-style-type:disc;">[0-9] 라고 쓰면 그 자리에 있는 한 글자가 0부터9까지의 수 중 아무 숫자 하나로 쓰여진 패턴을 가지고 있다는 뜻이다.</li>
                    <li style="list-style-type:disc;">[0-9a-zA-Z] 라고 쓰면 그 자리에 있는 한 글자가 0부터9까지의 수나 a-z까지의 알파벳 혹은 A-Z까지의 알파벳 중 아무 글자 하나로 쓰여진 패턴을 가지고 있다는 뜻이다.</li>
                    <li style="list-style-type:disc;">[^a-z] 라고 쓰면 그 자리에 있는 한 글자가 a-z까지의 알파벳으로는 절대 쓰이지 않은 패턴을 가지고 있다는 뜻이다.</li>
                    <li style="list-style-type:disc;">[\t\n\r] 라고 쓰면 그 자리에 있는 한 글자가 tab이나 엔터키(줄바꿈), 스페이스바(띄어쓰기) 중 아무 거나 하나로 쓰여진 패턴을 가지고 있다는 뜻이다.</li>
                    <li style="list-style-type:disc;">그래서 이 정규식 처리 모듈을 가지고 전화번호를 패턴화하면, 아래와 같이 만들어진다.</li>
