        <h3>2. 연산자</h3>
        <ul><ul>
            <li style="list-style-type:disc;">연산자는 데이터/값들에 대한 연산을 정의한다.</li>
            <ul><ul>
                <li style="list-style-type:circle;"><strong>자료형에 따라 연산자가 정의된다.</strong> 예를 들면 int 자료형과 int 자료형을 더하는 연산자와 float 자료형과 float 자료형을 더하는 연산자는 같은 +를 쓰지만, 컴퓨터는 int 자료형과 int 자료형을 더하는 +를 int 덧셈용 연산자라고 인식하고, float 자료형과 float 자료형을 더하는 +를 float 덧셈용 연산자라고 인식한다.</li>
                <li style="list-style-type:circle;">위의 예시에서 int 덧셈용 연산자가 +이고, float 자료형 덧셈용 연산자가 +인 것처럼, <strong>같은 연산자가 자료형에 따라 다른 동작을 하기도 한다.</strong></li>
                <li style="list-style-type:circle;"><strong>연산자는 하나 혹은 그 이상의 피연산자(연산 되어지는 것, operand)를 가질 수 있다.</strong> 예를 들어 – (-1)과 같은 경우에 맨 앞의 -는 (-1)이라는 1개의 피연산자를 가지고 있는 상황인 것이고, 1+2와 같은 경우의 +는 1과 2라는 2개의 피연산자를 가지고 있는 상황인 것이다.</li>

        <h3>4. 변수</h3>
        <ul><ul>
            <h4>A. 변수가 필요한 이유</h4>
            <ul><ul>
                <li style="list-style-type:disc;">동일한 값을 여러 번 사용할 필요가 있다.</li>
                <li style="list-style-type:disc;">중간 결과를 나타내기 위해서 사용한다.</li>
                <li style="list-style-type:disc;">수학에서 사용하는 변수와 유사하다.</li>
            </ul></ul>
            <h4>B. 정의</h4>
            <ul><ul>
                <li style="list-style-type:disc;"><strong>어떠한 값(value) 혹은 객체(object)에 대한 이름(alias)을 변수라고 한다.</strong></li>
                <li style="list-style-type:disc;"><strong>엄밀히 말하면, 값(value) 혹은 객체(object)를 저장할 수 있는 이름이 붙어있는 메모리 상의 공간을 변수라고 한다.</strong></li>
                <li style="list-style-type:disc;">변수에 값 혹은 객체를 대입(assignment)한 이후, 대입된 값 혹은 객체 대신에 변수의 이름을 사용할 수 있다.</li>
            </ul></ul>
            <h4>C. 변수 이름의 규칙</h4>
            <ul><ul>
                <li style="list-style-type:disc;">규칙</li>
                <ul><ul>
                    <li style="list-style-type:circle;">변수 이름의 첫 글자는 반드시 알파벳 문자 혹은 밑줄(_)이여야 한다.</li>
                    <li style="list-style-type:circle;">변수 이름의 나머지 글자는 문자, 밑줄, 숫자 모두 가능하다.</li>
                    <li style="list-style-type:circle;">컴퓨터는 변수의 대/소문자를 구별해서 다르게 인식한다. 예를 들면, pi, Pi, PI의 3가지 변수가 있을 때, 컴퓨터는 이 변수들을 서로 다른 값으로 인식한다.</li>
                    <li style="list-style-type:circle;">python keyword(python의 내부함수 및 연산자들, 예약어라고 부름)는 변수 이름으로 설정할 수 없다.</li>
                </ul></ul>
                <li style="list-style-type:disc;">예시</li>
                <ul><ul>
                    <li style="list-style-type:circle;">좋은 예: I, name_2_3</li>
                    <li style="list-style-type:circle;">나쁜 예: 2things, two words, my-name, >a1b2_3</li>
                </ul></ul>
                <li style="list-style-type:disc;">일반적으로 python에서의 변수이름은 할당되는 값들이 어떤 값들인지 잘 나타낼 수 있는 이름으로 설정한다. 또한 변수이름은 모두 소문자로 작성하고, 단어 사이에서는 밑줄을 사용한다.</li>
            </ul></ul>
            <h4>D. 변수 대입(assignment)</h4>
            <ul><ul>
                <li style="list-style-type:disc;">variable_name = expression의 꼴을 변수 대입문이라 정의한다.</li>
                <ul><ul>
                    <li style="list-style-type:circle;">의미: 오른쪽 expression의 결과값을 왼쪽 변수(할당된 저장소 공간)에 집어넣으라(대입하라)는 명령이다.</li>
                    <li style="list-style-type:circle;">주의: 할당할 변수의 이름은 항상 왼쪽에 위치하여야 하고, 할당될 변수 혹은 expression은 오른쪽에 위치하여야 한다.</li>
                </ul></ul>

            <h4>E. 변수 축약형</h4>
            <ul><ul>
                <li style="list-style-type:disc;">variable #= expression의 꼴을 변수 축약형이라 정의한다.</li>
                <ul><ul>
                    <li style="list-style-type:circle;">의미: variable = variable # expression의 축약형태로, <strong>변수의 값을 이용해 expression과 #연산을 하고, 다시 그 값을 변수에 대입하는 대입연산자를 변수 축약형이라고 한다.</strong></li>
                    <li style="list-style-type:circle;">이 변수 축약형이 실행되면, 가장 먼저 expression이 계산되고, 그 뒤에 변수와 expression이 계산된다.</li>
                    <li style="list-style-type:circle;">#의 자리에 들어갈 수 있는 연산자는 +, -, *, /, //, % 등이다.</li>
                </ul></ul>
                <li style="list-style-type:disc;">사용 예시</li>
                <ul><ul>
                    <li style="list-style-type:circle;">var += 3  ⇒  var = var + 3</li>
                    <li style="list-style-type:circle;">t *= 2 + 3  ⇒  t = t * (2+3)</li>
                </ul></ul>
            </ul></ul>
            <h4>F. 문자열과 변수</h4>
            <ul><ul>
                <li style="list-style-type:disc;">따옴표(‘ 또는 “)로 둘러싸여 있는 문자들을 문자열이라고 한다.</li>
                <li style="list-style-type:disc;">어떠한 값 또는 객체를 나타내는 이름을 문자열이라고 한다.</li>
                <li style="list-style-type:disc;">예제</li>
                <ul><ul>
                    <li style="list-style-type:circle;">n = 42  ⇒  n이라는 변수에 42 할당하기.</li>
                    <li style="list-style-type:circle;">var1 = n  ⇒  var1이라는 변수에 n이라는 변수의 값(=42) 할당하기.</li>
                    <li style="list-style-type:circle;">var2 = ‘n’ ⇒  var2이라는 변수에 ‘n’이라는 문자열 할당하기.</li>
                    <li style="list-style-type:circle;">var3 = n + 1  ⇒  var3라는 변수에 n이라는 변수의 값(=42)에 1을 더한 값(=43) 할당하기.</li>
                    <li style="list-style-type:circle;">var4 =’n’ + 1  ⇒  Error 발생(문자열과 숫자를 더하는 연산자는 정의되어 있지 않다.)</li>
                </ul></ul>
            </ul></ul>

            <li style="list-style-type:disc;">수학에서 인자와 동일한 역할을 하는 <strong>인자(argument)를 가질 수 있다.</strong></li>
            <li style="list-style-type:disc;">수학에서 함수값과 동일한 역할인 <strong>반환값(return value)을 가질 수 있다.</strong></li>
        </ul></ul>

        <h3>6. 표준 출력(standard output): print( )</h3>
        <ul><ul>
            <li style="list-style-type:disc;">컴퓨터의 메모리에 저장된 값을 출력매체를 통해 보여주도록 하는 함수를 표준 출력함수라고 하며, print( )의 꼴로 정의한다.</li>
            <li style="list-style-type:disc;">print( )의 역할</li>
            <ul><ul>
                <li style="list-style-type:circle;">인자에 있는 <strong>내용을 화면에 출력하고 줄을 바꾼다.</strong></li>
                <li style="list-style-type:circle;"><strong>인자가 여러 개 있는 경우, 여러 인자 사이에 ,를 찍어 입력해주도록 하며 이 경우 ,가 빈 칸(Space 1칸)으로 변경되어 출력된다.</strong></li>
                <li style="list-style-type:circle;"><strong>문자열의 빈 칸(Space 1칸)이 아닌 이상 그 외의 빈 칸(Space 1칸)은 특별한 의미를 가지지 않으며, 출력되지도 않는다.</strong></li>
            </ul></ul>

        <h3>7. 표준 입력(standard input): input( )</h3>
        <ul><ul>
            <li style="list-style-type:disc;">컴퓨터의 메모리에 입력매체를 통해 값을 저장하도록 하는 함수를 표준 입력 함수라고 하며, input( )의 꼴로 정의한다.</li>
            <li style="list-style-type:disc;">‘프로그램으로 들어가는 데이터 스트림(일반적으로 키보드 입력)’라고 표현하기도 한다.</li>
            <li style="list-style-type:disc;">input( )의 역할</li>
            <ul><ul>
                <li style="list-style-type:circle;">일반적으로 키보드로 입력받은 내용을 문자열로 반환하는 역할을 한다. </li>
                <li style="list-style-type:circle;"><strong>보통 키보드로 입력받은 내용을 문자열로 반환하여 변수로서 메모리에 저장하고 사용하기 위해서 variable_name = input( )의 꼴로 사용</strong>되는데, input( )의 꼴로 단독으로 쓰면 입력받은 값이 문자열로 반환된 후 저장되지 않고 바로 사라지기 때문이다.</li>
                <li style="list-style-type:circle;">함수( ) 안에 인자(변수 혹은 값이나 문자열)를 집어넣을 경우, 인자의 내용을 화면에 출력한다.</li>
