    <h2>ⅰ. 주석 (comment)</h2>
    <ul><ul>
        <h3>1. 주석의 필요성</h3>
        <ul><ul>
            <li style="list-style-type:disc;">주석은 다른 사람이나 스스로가  코드를 이해할 수 있도록(코드의 가독성을 높이기 위해서) 코드에 대한 설명을 해놓은 것이다.</li>
            <li style="list-style-type:disc;">함수가 정의된 후 코드블럭의 첫줄에 주석이 적혀있는 경우 함수의 역할을 설명하는 주석이라는 것이 약속되어 있다.</li>
            <img src="./Imgs for W6Note/necessity of comment 1.png" alt="" loading="lazy"  align="left" style="width:500px;margin-right:100%;">
            <img src="./Imgs for W6Note/necessity of comment 2.png" alt="" loading="lazy"  align="left" style="width:500px;margin-right:100%;">
        </ul></ul>
        <h3>2. 주석을 넣는 방법</h3>
        <ul><ul>
            <li style="list-style-type:disc;">#을 이용하면, #부터 그 줄의 끝까지는 주석으로 처리된다.</li>
            <li style="list-style-type:disc;">여러 줄의 주석을 다는 경우</li>
            <ul><ul>
                <li style="list-style-type:circle;">모든 줄에 #을 표시한다.</li>
                <li style="list-style-type:circle;">따옴표 3개를 사용하여(‘’’ ‘’’), 여러줄에 걸친 문자열을 사용한다.</li>
            </ul></ul>
        </ul></ul>
        <h3>3. 주석을 넣을 때의 주의점</h3>
        <ul><ul>
            <li style="list-style-type:disc;">주석은 코드에 대해 설명하는 것이므로, 코드와 일관되도록 적어주어야 한다.</li>
            <li style="list-style-type:disc;">주석을 잘못 넣으면, 오히려 코드를 이해하는 것이 더 어려워진다.</li>
            <li style="list-style-type:disc;">코드블럭 안에 주석을 넣을 때는 주석도 코드블럭 안에 있음을 표시하기 위해 띄어쓰기 4칸을 한 후에 #을 쓰거나 ‘’’을 써서 주석을 처리하도록 한다.</li>
        </ul></ul>
    </ul></ul>

    <h2>ⅲ. 자료구조</h2>
    <ul><ul>
        <h3>1. 자료구조: 자료를 구조화하고 저장하는 방식을 자료구조라고 한다.</h3>
        <ul><ul>
            <li style="list-style-type:disc;">하나의 변수에 하나 이상의 값 혹은 객체를 저장하기 위하여 주로 사용된다.</li>
            <li style="list-style-type:disc;">Python은 특성이 다른 여러 형태의 자료구조를 제공한다.</li>
            <ul><ul>
                <li style="list-style-type:circle;">리스트(list)</li>
                <li style="list-style-type:circle;">튜플(tuple)</li>
                <li style="list-style-type:circle;">사전(dict)</li>
                <li style="list-style-type:circle;">집합(set)</li>
            </ul></ul>
        </ul></ul>
        <h3>2. 시퀀스(sequence)</h3>
        <ul><ul>
            <li style="list-style-type:disc;">시퀀스는 python의 자료구조의 종류의 하나이다.</li>
            <ul><ul>
                <li style="list-style-type:circle;">변경가능한 시퀀스(mutable): list</li>
                <li style="list-style-type:circle;">변경 불가능한 시퀀스(immutable): tuple, str</li>
            </ul></ul>
            <li style="list-style-type:disc;">시퀀스의 공통적인 특징</li>
            <ul><ul>
                <li style="list-style-type:circle;">유한한 순서가 있는 집합으로, 음이 아닌 정수로 그 순서(index)를 나타낸다.</li>
                <li style="list-style-type:circle;">슬라이싱(slicing, 시퀀스의 부분집합)이 제공된다.</li>
                <li style="list-style-type:circle;">유사한 형태의 여러 연산자 혹은 함수가 제공된다.</li>
            </ul></ul>
        </ul></ul>
        <h3>3. 리스트(list)</h3>
        <ul><ul>
            <h4>A. 정의: mutable sequences, typically used to store collections of homogeneous items</h4>
            <ul><ul>
                <li style="list-style-type:disc;">mutable: 원소의 추가/삭제 혹은 값이 변경될 수 있음.</li>
                <li style="list-style-type:disc;">원소의 자료형: 임의의 자료형의 원소를 저장할 수 있음.</li>
            </ul></ul>
            <h4>B. 인덱스(index): 음이 아닌 정수로 원소의 순서를 나타낸다.</h4>
            <h4>C. 표기법: []를 사용하고 원소는 ,로 구분한다.</h4>
            <ul><ul>
                <img src="./Imgs for W6Note/orthography of list.png" alt="" loading="lazy"  align="left" style="width:500px;margin-right:100%;">
                <ul><ul>
                    <li style="list-style-type:disc;">빈 리스트도 안에 데이터만 없을 뿐 리스트 자료형이 맞다.</li>
                </ul></ul>
            </ul></ul>
            <h4>D. list의 원소 접근</h4>
            <ul><ul>
                <li style="list-style-type:disc;">리스트의 원소는 0부터 시작하는 인덱스로 접근할 수 있다.</li>
                <li style="list-style-type:disc;">음수가 사용되면, 가장 마지막 원소부터 역순으로 접근한다.</li>
                <img src="./Imgs for W6Note/access to list item 1.png" alt="" loading="lazy"  align="left" style="width:500px;margin-right:100%;">
                <li style="list-style-type:disc;">리스트는 namespace에서 list로써 저장되어 있으며, 그 list가 불려지면, 그 list를 저장하고 있는 다른 메모리 공간에 가서 list를 불러온다. list가 저장된 다른 메모리 공간에 list는 인덱스가 0부터 오름차순으로 데이터에 붙어서 정렬되어 있다.</li>
                <img src="./Imgs for W6Note/access to list item 2.png" alt="" loading="lazy"  align="left" style="width:500px;margin-right:100%;">
            </ul></ul>
            <h4>E. list의 여러 원소들 접근: 슬라이싱(slicing)</h4>
            <ul><ul>
                <li style="list-style-type:disc;"><strong>name_of_list[start: end]</strong></li>
                <ul><ul>
                    <li style="list-style-type:circle;">name_of_list의 [start: end]에 위치한 원소들을 가지는 리스트를 생성한다. start번째 인덱스를 가진 데이터부터 end번째 인덱스를 가진 데이터 전까지 가져온다.</li>
                    <li style="list-style-type:circle;">start: 생략되면 첫 원소부터 가져온다.</li>
                    <li style="list-style-type:circle;">end: 생략되면 마지막 원소까지 가져온다.</li>
                    <li style="list-style-type:circle;">start와 end 모두 생략되면, 해당하는 리스트 전체를 다 가져온다.</li>

            </ul></ul>
            <h4>F. list에 대한 연산(이 연산 중 대부분은 다른 시퀀스(str, tuple)에도 적용된다.)</h4>
            <ul><ul>

                    <li style="list-style-type:disc;">not in: not in 좌측의 데이터가 not in 우측의 시퀀스 안에 없는가를 연산하는 불연산자. 없으면 True, 있으면 False를 뱉는다.</li>

        <h3>4. range( ) 함수</h3>
        <ul><ul>
            <li style="list-style-type:disc;">Range( ) 함수</li>
            <ul><ul>
                <li style="list-style-type:circle;">range(stop): immutable sequence(불변의 시퀀스) I (0&#60;=i&#60;stop) 즉, 0부터 stop에 해당하는 숫자까지 i값만큼의 간격으로 나누어 한줄로 저장하는 함수. 0은 무조건 포함하고, 0에 i를 더해가면서 stop보다 작은 숫자이면 0 뒤에 이어서 저장하도록 하는 함수이다. 따라서 stop 값은 포함되지 않는다. 보통 i는 1이다.</li>
                <li style="list-style-type:circle;">range(start, stop, step): immutable sequence I (start&#60;=i&#60;stop, i+=step) 즉, 0부터 stop에 해당하는 숫자까지 step값만큼의 간격으로 나누어 한줄로 저장하는 함수. 0은 무조건 리스트에 포함하고, 0에 step 값을 더해가면서 stop보다 작은 숫자이면 0 뒤에 이어서 저장하도록 하는 함수이다. 따라서 stop 값은 리스트에 포함되지 않는다.</li>
                <ul><ul>
                    <img src="./Imgs for W6Note/range function.png" alt="" loading="lazy"  align="left" style="width:500px;margin-right:100%;">
                </ul></ul>
