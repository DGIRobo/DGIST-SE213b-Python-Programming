<h1>Ⅸ. 객체지향 프로그래밍 소개</h1>
<ul><ul>
    <h2>ⅰ. 객체지향 프로그래밍 (Object-Oriented Programming)</h2>
    <ul><ul>
        <h3></h3>
        <ul><ul>
            <h4>A. 객체지향 프로그래밍이란?</h4>
            <ul><ul>
                <li style="list-style-type:disc;"><strong>절차지향 프로그래밍 (Procedural Programming)</strong></li>
                <ul><ul>
                    <li style="list-style-type:circle;">프로시저 (Procedure: 프로그램 언어에 따라 서브루틴, 함수, 메소드 등으로도 불림) 호출을 사용하는 프로그래밍 패러다임을 절차지향 프로그래밍이라고 한다.</li>
                    <li style="list-style-type:circle;">모듈화를 통해 작업하고자 하는 독립적인 범위(scope)를 만들고, 인자와 반환값을 통해 프로시저와 데이터를 교환한다.</li>
                    <li style="list-style-type:circle;">어떠한 ‘작업’을 수행하는 것에 초점을 맞추어 프로그램을 개발하는 것이다.</li>
                </ul></ul>
                <li style="list-style-type:disc;"><strong>객체지향 프로그래밍 (Objected-Oriented Programming, OOP)</strong></li>
                <ul><ul>
                    <li style="list-style-type:circle;">절차지향 프로그래밍이 프로시저만 분리하여 프로그래밍을 하는 것이었기에 자료와는 분리되어 있다. 따라서 한가지 예로 add라는 함수가 있다면 이게 문자열을 연산하는 건지, 숫자를 연산하는건지, 또 다른 어떤 자료형을 연산하는 건지 알 수가 없다. 이 한계점 때문에 자료형과 그 자료형을 연산하는 연산(함수)을 함께(객체로) 정리하고자 하는 패러다임이 나왔다.</li>
                    <li style="list-style-type:circle;">데이터와 프로시저를 결합한 ‘객체’라는 개념에 기반한 프로그래밍 패러다임을 객체지향 프로그래밍이라고 한다.</li>
                    <li style="list-style-type:circle;">작업 대상이 되는 ‘자료’(Data)와 ‘처리방법’을 동시에 설계</li>
                </ul></ul>
                <li style="list-style-type:disc;">참고: 객체지향과 다른 ‘함수형 프로그래밍’도 있다.</li>
            </ul></ul>
            <h4>B. 왜 객체지향 프로그래밍을 사용하는가?</h4>
            <ul><ul>
                <li style="list-style-type:disc;">더 높은 수준의 추상화(abstract)를 제공하기 위한 방향으로 프로그래밍 언어가 진화하였다. (Machine language -> Assembly language -> Procedural languages C or Fotran -> Object-oriented languages C++ or Java, python -> Declarative or functional(함수 중심적 프로그래밍-수학적으로 증명하기에 용이할 것이다.) languages Lisp, Haskell, ML, prolog, Scala)</li>
                <li style="list-style-type:disc;">즉, 어떤 실체가 있거나 없는 개념을 무엇이든지 구현할 수 있도록 하는 방향으로 프로그래밍 언어가 진화하였다.</li>
                <li style="list-style-type:disc;"><strong>목적</strong></li>
                <ul><ul>
                    <li style="list-style-type:circle;">유지보수의 용이성: 데이터와 그 데이터에 대한 독립적인 함수가 같이 정리되어 있어서 구성이 쉽다. 따라서 유지보수가 용이하다.</li>
                    <li style="list-style-type:circle;">코드의 재사용성이 향상된다.: 함수 뿐만이 아니라 자료구조도 함께 모듈화했으므로 코드, 디자인 방법 등을 재사용할 수 있다.</li>
                    <li style="list-style-type:circle;">확장성 제공: 코드에 대한 요구사항이 변경되었을 때, 기능의 추가/변경이 용이하다.</li>
                </ul></ul>
            </ul></ul>
            <h4>C. 객체지향 프로그래밍을 지원하는 언어</h4>

                    <li style="list-style-type:circle;">ruby</li>
                    <li style="list-style-type:circle;">C#</li>
                    <li style="list-style-type:circle;">Objective-C</li>

                <li style="list-style-type:disc;">현재 널리 사용되는 대부분의 언어에서 객체지향 프로그래밍의 개념을 지원하고 있다.</li>

        <h3>2. 파이썬과 객체지향 프로그래밍</h3>
        <ul><ul>
            <h4>A. Python의 객체지향 프로그래밍 지원</h4>
            <ul><ul>
                <li style="list-style-type:disc;">python은 객체지향을 염두에 두고 설계된 언어이다.</li>
                <ul><ul>
                    <li style="list-style-type:circle;">간단한 문법으로 대부분의 객체지향 개념을 지원한다.</li>
                    <li style="list-style-type:circle;">엄격한 문법으로 객체지향을 지원하는 C++, Java와는 차이가 있다.</li>
                </ul></ul>
                <li style="list-style-type:disc;">python에서는 모든 것이 객체로 구현되어 있다. 따라서 기본자료형, 자료구조, 함수가 모두 객체지향적으로 구현되어 있어서 한 자료에 대해 그 자료만을 위한 연산이 짝으로 존재한다.</li>
                <li style="list-style-type:disc;">클래스(class) vs 인스턴스(instance) 혹은 객체(object)</li>
                <ul><ul>
                    <li style="list-style-type:circle;">클래스: 어떤 객체를 집어넣을 때, 컴퓨터가 그것을 인식할 수 있도록 정의한 것으로 자료형이라 부르기도 한다. 쉽게 말해, 다양한 책상다리의 구조를 컴퓨터가 바로 인식하는 것은 불가능하므로, 다리와 책상 그리고 재질 등을 컴퓨터에 미리 알려주고 어떤 책상다리의 구조가 입력되었을 때, 이를 인식하도록 만드는 작업을 해야 한다. 이때 미리 알려주는 작업이 클래스를 만드는 작업인 것이다. 어떤 객체의 형태(type)를 정의한 것이 class인 것이다.</li>
                    <ul><ul>
                        <p>ex. int, float</p>
                    </ul></ul>
                    <li style="list-style-type:circle;">인스턴스 (instance)/ 객체 (object): 특정 클래스에 값이 주어진 실체 </li>
                    <ul><ul>
                        <p>ex. 42, 1024, 3.14</p>
                    </ul></ul>
                </ul></ul>
            </ul></ul>
            <h4>B. Class (자료형을 나타내기 위한 설계도)의 구성요소</h4>
            <ul><ul>
                <li style="list-style-type:disc;">클래스의 속성 (attributes)</li>
                <ul><ul>
                    <li style="list-style-type:circle;">데이터 속성 (data attributes): Class를 만든다는 것이 자동차설계도를 만드는 것이라 할 때, 차문, 유리, 차체 등에 해당한다. Class의 변수를 말한다.</li>
                    <ul><ul>
                        <p>▶ member variable, field, attributes, properties, characteristics</p>
                    </ul></ul>
                    <li style="list-style-type:circle;">메소드(methods): Class를 자동차 설계도를 만드는 것이라 할 때, 데이터 속성을 가지고 할 수 있는 기능들을 정의한 것을 의미한다. 예를 들면, 엑셀을 밟아 연료를 분사시킬수 있다는 것이나 버튼을 눌러 창문을 내릴 수 있다는 것이 이에 해당한다. Class의 데이터 속성을 받아 처리하는 함수를 말한다.</li>
                    <ul><ul>
                        <p>▶ member functions, methods, responsibilities</p>
                    </ul></ul>
                </ul></ul>
                <li style="list-style-type:disc;">클래스 & 인스턴스 변수 (class & instance variable)</li>
                <ul><ul>
                    <li style="list-style-type:circle;">클래스 변수: Class를 만드는 것이 자동차 설계도를 만드는 것이라 할 때, 자동차의 상표와 같이 Class(자동차 설계도)를 보고 만든 모든 자동차(인스턴스 or 객체)들에 공유되는 변수를 말한다. 즉, Class 내에 정의되어 있기 때문에 그 Class를 통해 만든 인스턴스(or 객체)들이 공통적으로 가지고 있는 변수를 말한다. 따라서, Class를 만들 때, 이 Class를 사용하는 모든 인스턴스가 공통적으로 가져야 할 변수를 만들어야 한다면 클래스 변수로 선언한다.</li>
                    <li style="list-style-type:circle;">인스턴스 변수: Class를 만드는 것이 자동차 설계도를 만드는 것이라 할 때, 자동차의 번호판처럼 Class(자동차 설계도)를 보고 만든 모든 자동차(인스턴스 or 객체)들의 서로 다른 변수를 말한다. 즉, 클래스의 인스턴스 별로 독립적인 변수이다. 따라서, Class를 만들 때, 인스턴스 변수의 자리는 만들어주되, 인스턴스 변수값 할당은 하지 않는다. 인스턴스 변수는 인스턴스에 나타내져 있어야 한다.</li>
                </ul></ul>
                <li style="list-style-type:disc;">클래스 & 인스턴스 함수 (class & instance function)</li>
                <ul><ul>
                    <li style="list-style-type:circle;">클래스 함수: Class 내에 정의되어 있는 함수로, Class 내부의 클래스 변수를 확인하거나 수정할 수 있는 함수이다. Class 내에 정의되어 있는 함수이기에 이 Class를 공유하는 모든 인스턴스는 당연히 이 함수를 공유(사용가능하다는 것)한다. 당연히 Class 내에 없는 인스턴스 변수에는 접근 불가능하다.</li>
                    <li style="list-style-type:circle;">인스턴스 함수: Class 내에 정의되어 있는 함수로, 같은 Class를 공유하는 모든 인스턴스들 각각에서 자신의 인스턴스 변수를 접근하는 기능을 제공하는 함수이다. 첫 번째 인자의 이름은 어떤 것으로 해도 되지만, 관례적으로 self(인스턴스 자신을 의미함.)를 사용한다.</li>
                </ul></ul>
            </ul></ul>
            <h4>C. 자료형과 식별자를 반환하는 함수:  type( ), id( )</h4>
            <ul><ul>
                <li style="list-style-type:disc;">type( ): 어떠한 값 혹은 변수에 저장된 값의 자료형을 반환하는 함수이다.</li>
                <li style="list-style-type:disc;">id( ): 객체의 식별자(객체들을 구별하기 위한 unique한 값이며, 값 자체는 중요하지 않다.)를 반환하는 함수이다. (다른 언어에서는 보통 메모리 주소를 반환하지만, 파이썬은 그렇지 않으므로 신경쓰지 말자.)</li>
                <li style="list-style-type:disc;">같은 Class를 사용하는 서로 다른 인스턴스에 type함수를 쓰면, 같은 값이 나오고 id함수를 쓰면 다른 값이 나온다. 따라서 이름이 다른 인스턴스 2개가 주어졌을 때, 2개가 같은 것을 가리키고 있는지 확인하려면, id 함수를 써보면 된다. 2개가 같은 값이면 같은 인스턴스를 가리키는 2개의 변수명인 것이고, 다른 값이면 서로 다른 인스턴스를 가리키는 2개의 변수명인 것이다.</li>
                <img src="./Imgs for W12Note/type, id.png" alt="" loading="lazy"  align="left" style="width:500px;margin-right:100%;">
            </ul></ul>
            <h4>D. 식별자 혹은 객체의 값의 동일 여부를 확인하는 예약어: is, is not, ==, !=</h4>
            <ul><ul>
                <li style="list-style-type:disc;">is, is not: 동일한 객체(인스턴스)를 가르키는지 확인하는 예약어이다. 즉, 식별자가 동일한지 판단하는 연산이다.</li>
                <li style="list-style-type:disc;">==, !=: 객체가 저장하고 있는 내용이 동일한지를 판단하는 연산이다.</li>
                <li style="list-style-type:disc;">(Advanced) 사용자가 만든 클래스의 경우,</li>
                <ul><ul>
                    <li style="list-style-type:circle;">== 연산자 계산을 하려면, Class 내부에 인자로 넘겨진 객체의 내용과 현재 객체 내용의 동일 여부를 검사해서 True 혹은 False를 반환하는 특수한 함수(__eq__())를 정의해주어야 한다. </li>
                    <li style="list-style-type:circle;">Class에 이 함수를 만들지 않았을 경우에는 is 연산자의 결과가 == 연산자의 결과로 사용된다.</li>
                </ul></ul>

            <h4>C. 클래스가 제공하는 연산: 속성 참조와 객체화</h4>
            <ul><ul>
                <li style="list-style-type:disc;">속성 참조 (attribute reference): 클래스 내에 있는 클래스 변수/함수 및 인스턴스 변수/함수를 접근하는 방법을 말한다.</li>
                <ul><ul>
                    <li style="list-style-type:circle;">클래스 변수/함수</li>
                    <ul><ul>
                        <p>▶ ClassName.variable_name</p>
                        <p>▶ ClassName.function_name()</p>
                    </ul></ul>
                    <li style="list-style-type:circle;">인스턴스 변수/함수</li>
                    <ul><ul>
                        <p>▶instance_name.variable_name</p>
                        <p>▶instance_name.function_name()</p>
                    </ul></ul>
                </ul></ul>
                <li style="list-style-type:disc;">객체화 (instantiation): 클래스 파일을 만들고, 다른 파일에서 이 클래스 파일을 부르면(import) 이 클래스는 객체화된다. 그러면 이 클래스를 부른 파일은, 클래스 이름을 함수의 형태로 사용하여, 클래스를 객체로서 사용할 수 있게 된다.</li>
                <ul><ul>
                    <p>▶ ClassName()</p>
                </ul></ul>
            </ul></ul>

        <h3>4. Class 변수/함수의 사용</h3>
        <ul><ul>
            <h4>A. 클래스 변수/함수의 정의와 사용</h4>
            <ul><ul>
                <li style="list-style-type:disc;">클래스가 선언되면, 클래스의 정보(어떤 변수, 함수를 가지는지)가 전역 네임스페이스에 저장된다.</li>

        <h3>5. 인스턴스 변수/함수의 사용</h3>
        <ul><ul>
            <h4>A. 클래스의 인스턴스화: 세부적인 사항</h4>
            <ul><ul>
                <li style="list-style-type:disc;">클래스를 인스턴스화 할 때, python이 하는 작업은 다음과 같다.</li>
                <ul><ul>
                    <p>⑴ 인스턴스를 저장할 공간(일종의 네임스페이스)를 만든다.</p>
                    <p>⑵ __init__(self, …) 함수를 호출한다.</p>
                    <ul><ul>
                        <p>① 이때, 첫 번째 인자인 self는 ⑴에서 만든 공간을 가리킨다.</p>
                        <p>② self 이외의 다른 인자들은 함수 호출과 같은 규칙으로 __init__()의 매개변수로 전달된다.</p>
                    </ul></ul>
                    <p>⑶ 새로 만들어진 공간(즉 self가 가르키는 공간)이 반환값처럼 반환된다.</p>
                </ul></ul>
                <li style="list-style-type:disc;">__init__() 함수의 특성</li>
                <ul><ul>
                    <li style="list-style-type:circle;">이름은 바꿀 수 없다. 무조건 __init__()이라고 써야 한다.</li>
                    <li style="list-style-type:circle;">반환값을 가질 수 없다.</li>
                    <li style="list-style-type:circle;">참조값을 전달받는 첫 번째 매개 변수의 이름은 관례적으로 self를 사용한다.</li>
                </ul></ul>
            </ul></ul>
            <h4>B. 예시: 클래스의 인스턴스화</h4>

            <h4>G. Class에 직접 Operation 연산자를 써서 연산을 진행하도록 하는 방법(Binary Arithmetic Operations)</h4>
            <ul><ul>
                <img src="./Imgs for W12Note/binary arithmetic operations.png" alt="" loading="lazy"  align="left" style="width:550px;margin-right:100%;">
                <ul><ul>
                    <li style="list-style-type:disc;">왼쪽에서는 c1 객체 안의 counter값, c2 객체 안의 counter값을 더하는 add 함수(Class 함수이다.)를 정의했다. 그래서 마지막 줄 바로 전 줄에 보면, 이 함수를 쓰기 위해서 MyCounter.add(c1, c2)와 같은 함수 호출문을 작성하였다.</li>
                    <li style="list-style-type:disc;">그런데, Myconter의 인스턴스화(객체화)에 따라 생성된 객체 내에는 어짜피 변수가 counter 밖에 없다. 그러면 그냥 c1 + c2로 쓰면, c1 객체 안의 counter값, c2 객체 안의 counter값을 더하는 작업이 진행되도록 만들고 싶다고 생각이 든다. 그렇다고 왼쪽 코드 마지막 줄처럼 c1 + c2만 추가해서는 연산이 되지 않는다.</li>
                    <li style="list-style-type:disc;">이때, 사용되는 것이 __add__() 함수이다. 오른쪽 코드처럼 __add__(obj1 + obj2)라는 함수를 정의하게 되면, MyCounter.__add__(c1, c2)로 써도 c1 객체 안의 counter값, c2 객체 안의 counter값을 더하는 작업이 일어나며, c1.__add__(c2)라고 써도 c1 객체 안의 counter값, c2 객체 안의 counter값을 더하는 작업이 일어난다. 또한, c1 + c2라고 써도 c1 객체 안의 counter값, c2 객체 안의 counter값을 더하는 작업이 일어난다. python의 interpreter가 +를 __add__() 함수로 변환하여 받아들이기 때문이다.</li>
                    <li style="list-style-type:disc;">이 __add__() 외에도 python의 interpreter에는 특수 기호로 표현된 연산자들을 의미하는 함수들이 정의되어 있다. 보통 그런 함수들은 __function__()와 같은 형식으로 정의되어 있으므로, 내가 만든 Class에서 그와 같은 기능을 제공해주고 싶다면 그런 함수들을 찾아 정의해주어야 한다. 이 내용은 <a href="https://docs.python.org/3.7/reference/datamodel.html?highlight=add#object.add">https://docs.python.org/3.7/reference/datamodel.html?highlight=add#object.add</a>에 나와있다.</li>
                </ul></ul>
            </ul></ul>
            <h4>H. 사용하는 이름 목록을 반환하는 함수: dir()</h4>
            <ul><ul>
                <img src="./Imgs for W12Note/dir function.png" alt="" loading="lazy"  align="left" style="width:550px;margin-right:100%;">
                <ul><ul>
                    <li style="list-style-type:disc;">dir() 함수: dir(object)의 형식으로 사용하는 함수로, object를 대상으로 사용할 수 있는 함수들의 이름과 object에서 호출하거나 할당할 수 있는 변수들의 이름을 리스트로 만들어 반환하는 함수이다.</li>
                    <li style="list-style-type:disc;">dict(object)에서 object로 class name을 집어넣으면, 그 class의 class 함수, class 변수, 인스턴스 함수가 나온다.</li>
                    <li style="list-style-type:disc;">dict(object)에서 object로 instance name을 집어넣으면, 그 instance의 class 함수, class 변수, 인스턴스 함수, 인스턴스 변수가 나온다.</li>
                    <li style="list-style-type:disc;">남이 만들어 놓은 객체를 내가 사용하고자 할 때, 어떻게 사용할 수 있는지 전혀 모르므로 그 객체를 대상으로 할 수 있는 작업들을 모두 찾아보아야 한다. 이때, 이 작업들의 이름을 리스트로 만들어 리턴해주는 dir() 함수를 사용하는 것이다.</li>
                    <li style="list-style-type:disc;">그런데, dir( ) 함수를 사용하면 class에 정의된 것보다 더 많은 함수, 변수들이 나오는데, class를 정의하면 거기에 정의한 함수, 변수에 python이 기본적으로 지원하는 함수들, 변수들이 추가되기 때문이다.</li>
