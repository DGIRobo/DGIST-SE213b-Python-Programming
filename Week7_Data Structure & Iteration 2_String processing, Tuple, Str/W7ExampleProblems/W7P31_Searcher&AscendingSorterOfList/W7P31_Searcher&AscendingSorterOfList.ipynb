### <span style='font-family:Times New Roman'> search(alphabet,word_list) 함수
<br>

> ##### <li> <span style='font-family:Times New Roman'> 함수의 동작 </li>
```Bash
단어들이 리스트로 저장되어 있는 word_list를 매개변수로 받아서,
단어의 첫 글자가 또다른 매개변수 alphabet과 동일한 경우
결과 리스트에 저장하고 출력 후 종료
```

> ##### <li> <span style='font-family:Times New Roman'> 매개변수 </li>
```Bash
alphabet: (str) 알파벳 1글자
word_list: (list) 단어들이 저장되어 있는 리스트
```

### <span style='font-family:Times New Roman'> dictionary(word_list) 함수
<br>

> ##### <li> <span style='font-family:Times New Roman'> 함수의 동작 </li>
```Bash
단어들이 리스트로 저장되어 있는 word_list를 매개변수로 받아서,
알파벳의 오른 차순으로 정렬한 뒤 출력 후 종료
```

> ##### <li> <span style='font-family:Times New Roman'> 매개변수 </li>
```Bash
word_list: (list) 단어들이 저장되어 있는 리스트
```

### <span style='font-family:Georgia'> Function Running Example
```Bash
['gray', 'goat']
['apple', 'bear', 'book', 'cap', 'doctor', 'goat', 'gray', 'pink', 'purple']
```

````python
# Answer of Problem 
def search(alphabet,word_list):
    search_list = []
    for val in word_list:
        if alphabet == val[0]:
            search_list.append(val)
    print(search_list)

def dictionary(word_list):
    word_list.sort()
    print(word_list)

example = ['cap','gray','bear','book','pink','purple','apple','goat','doctor']
search('g',example)
dictionary(example)
````
