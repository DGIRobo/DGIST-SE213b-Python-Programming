            <h4>A. 함수란?</h4>
            <ul><ul>
                <li style="list-style-type:disc;"><strong>함수(function or subroutine/procedure)란?</strong></li>
                <ul><ul>
                    <li style="list-style-type:circle;">특정 값을 입력 받아 명령문들의 집합을 실행하여 결과값을 출력하는 역할을 하는 부분(코드)이다.</li>

                </ul></ul>

                <ul><ul>
                    <li style="list-style-type:circle;">공통된 부분의 코드를 재사용할 수 있도록 하기 위해서 사용한다.</li>

                    <li style="list-style-type:circle;">반복과 오류를 줄이기 위해 사용한다.</li>
                    <li style="list-style-type:circle;">추상화(프로그래밍 코드에 추상적 역할을 할당하는 것)를 제공하기 위해 사용한다.</li>

            <h4>F. 매개변수의 기본값 지정</h4>
            <ul><ul>
                <li style="list-style-type:disc;">함수를 정의할 때, 매개변수의 기본값(default value)을 지정할 수 있다. </li>
                <ul><ul>
                    <li style="list-style-type:circle;">기본값(default value)을 지정하는 방법은 매개변수 이름 뒤에 = default_value의 형태로 기본값(default value)을 덧붙이는 것이다.</li>
                    <li style="list-style-type:circle;">이렇게 매개변수의 기본값을 지정하면, 함수를 호출할 때, 인자를 넘겨주지 않은 경우에는 기본값을 사용하게 된다.</li>
                </ul></ul>
                <li style="list-style-type:disc;">다만, 함수를 정의할 때, 기본값이 있는 매개변수들은 반드시 기본값이 없는 매개변수들 뒤에서 값을 받을 수 있도록 정의해주어야 한다.</li>
                <li style="list-style-type:disc;">기본값이 있는 매개변수들을 여러 개 정의하고 인자를 매개변수의 개수만큼 주지 않으면, 앞쪽에 있는 기본값이 있는 매개변수는 인자를 받아 사용하고 뒤쪽에 있는 기본값이 있는 매개변수에 인자를 넘겨주지는 않는다. 그래서 뒤쪽에 있는 인자를 못받은 매개변수들은 기본값을 사용하게 된다.</li>

        <h3>3. 네임스페이스와 지역 변수</h3>
        <ul><ul>
            <h4>A. 네임스페이스(namespace): 변수, 함수 등이 정의되어 있는 공간을 말한다.</h4>
            <h4>B. Python에서의 네임스페이스의 종류 및 생성/소멸</h4>
            <ul><ul>
                <li style="list-style-type:disc;">전역 네임스페이스 (global namespace): 프로그램이 실행될 때 생성되어 내장 함수와 전역 변수(함수 외부에서 정의된 변수)를 저장한다.</li>
                <li style="list-style-type:disc;">지역 네임스페이스 (local namespace): 함수가 호출될 때 생성되어 함수가 받는 매개 변수와 함수 안에서 정의된 변수(지역변수)를 저장한다. 그리고 함수의 실행이 끝나면 소멸된다.</li>
            </ul></ul>
            <h4>C. 변수의 범위(scope)</h4>
            <ul><ul>
                <li style="list-style-type:disc;">변수의 범위(scope): python에서 네임 스페이스, 즉 변수 혹은 함수가 접근가능한 영역을 변수의 범위(scope)라고 부른다.</li>
                <li style="list-style-type:disc;">변수 혹은 함수가 사용될 때, 컴퓨터는</li>
                <ul><ul>
                    <li style="list-style-type:circle;">우선 지역 네임스페이스에서 해당하는 변수 혹은 함수를 찾는다.</li>
                    <li style="list-style-type:circle;">지역 네임스페이스에 그 변수 혹은 함수가 없으면, 그 다음으로 전역 네임스페이스에서 그 변수 혹은 함수를 찾는다.</li>
                    <li style="list-style-type:circle;">전역 네임스페이스에서도 그 변수 혹은 함수를 찾을 수 없다면 NameError가 발생한다.</li>
