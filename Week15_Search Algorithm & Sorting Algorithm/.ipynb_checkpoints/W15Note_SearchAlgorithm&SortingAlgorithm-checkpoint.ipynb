<h1>ⅩⅡ. 검색 알고리즘(Search algorithm)과 정렬 알고리즘(Sort algorithm)</h1>
<ul><ul>
    <h2>ⅰ. 알고리즘이란</h2>
    <ul><ul>
        <h3>1. 알고리즘에 대한 소개</h3>
        <ul><ul>
            <li style="list-style-type:disc;"><strong>알고리즘</strong></li>
            <ul><ul>
                <li style="list-style-type:circle;">특정한 유형의 문제를 해결하기 위한 체계적인 방법</li>
                <li style="list-style-type:circle;">어떠한 형태의 입력에 대해서도 정확한 답을 찾는 방법</li>
                <li style="list-style-type:circle;">예시: 10진수의 사칙연산, 검색, 정렬,….</li>
            </ul></ul>
            <li style="list-style-type:disc;"><strong>알고리즘 복잡도</strong>: 알고리즘의 좋고 나쁨을 따지는 보편적인 기준.</li>
            <ul><ul>
                <li style="list-style-type:circle;">시간 복잡도: 얼마나 빨리 실행되는지?</li>
                <li style="list-style-type:circle;">공간 복잡도: 얼마나 많은 추가 공간을 필요로 하나?</li>
            </ul></ul>
        </ul></ul>
    </ul></ul>
    <br>
    <hr>
    <h2>ⅱ. 검색(탐색) 알고리즘</h2>
    <ul><ul>
        <h3>1. 검색(searching): 문제의 정의</h3>
        <ul><ul>
            <li style="list-style-type:disc;">리스트에서 찾고자 하는 원소가 어디에 있는지, 즉 인덱스를 찾는 것을 검색(searching)이라 한다.</li>
            <ul><ul>
                <li style="list-style-type:circle;">예시:</li>

                    </ul></ul>
                </ul></ul>
            </ul></ul>
        </ul></ul>
        <br>
        <hr>
        <h3>2. 순차 검색(sequential search)</h3>
        <ul><ul>
            <li style="list-style-type:disc;">원하는 값을 찾을 때까지, 리스트의 맨 앞에서부터 원소를 하나씩 빼서 살펴보는 알고리즘을 순차 검색(sequential search)라고 한다.</li>
            <li style="list-style-type:disc;">데이터의 구조를 모를때 가장 간단하게 사용할 수 있는 방법의 검색 알고리즘이며, 매우 구현하기 쉽고 간단하다.</li>
            <ul><ul>

        <h3>3. 이진 검색(binary search)</h3>
        <ul><ul>
            <li style="list-style-type:disc;">그런데, 순차 검색을 사용하면 데이터 개수가 많을 때, 검색 속도가 매우 느리다는 단점이 생긴다. 그래서 이진 검색(binary search)라는 알고리즘이 등장했다.</li>
            <li style="list-style-type:disc;"><strong>우선 이진 검색(binary search)을 사용하기 위해서는 검색할 리스트가 정렬되어 있어야 한다. 즉, 이진 검색(binary search)은 리스트가 정렬되어 있는 경우, 더 효율적인 검색을 하기 위한 방법인 것이다.</strong></li>
            <li style="list-style-type:disc;">이진 검색의 알고리즘: 리스트가 (오름차순으로) 정렬되어 있을 때, 원하는 원소를 찾기 위해</li>
            <ul><ul>
                <li style="list-style-type:circle;">처음에는 리스트 전체를 검색을 하는 범위로 설정한다.</li>
                <li style="list-style-type:circle;">검색을 하는 범위의 가운데 있는 수가 원하는 원소인지 비교한다.</li>
                <ul><ul>
                    <text>▶ 검색을 하는 수가 맞으면, 검색을 완료한다.<br></text>
                    <text>▶ 가운데 있는 수가 원하는 수보다 크면, 처음부터 가운데 하나 전까지를 검색하는 범위로 설정한 후 반복한다.<br></text>
                    <text>▶ 가운데 있는 수가 원하는 수보다 작다면, 가운데 하나 다음부터 끝까지를 검색하는 범위로 설정한 후 반복한다.</text>
                </ul></ul>
            </ul></ul>
            <li style="list-style-type:disc;">이 이진 검색(Binary search) 알고리즘은 우리가 사전을 찾는 방법을 알고리즘화 한 것이다. 사전을 펼쳐서 단어의 순서보다 앞인지 뒤인지 계속 판별하는 것으로 최종적으로 단어를 찾아내는 것을 알고리즘화 한 것이다.</li>
            <li style="list-style-type:disc;">예시: Find the item 9 in sorted list [2, 5, 6, 9, 12, 15, 16, 19, 24, 27, 30]</li>
            <ul><ul>
                <img src="./Imgs for W15Note/concept of binary search.png" alt="" loading="lazy" align="left" style="width:500px;margin-right:100%;">
            </ul></ul>

                <img src="./Imgs for W15Note/implement of binary search.png" alt="" loading="lazy" align="left" style="width:500px;margin-right:100%;">

    <h2>ⅲ. 알고리즘 복잡도(Algorithm complexity)</h2>
    <ul><ul>
        <h3>1. 알고리즘 복잡도(Algorithm complexity)</h3>
        <ul><ul>
            <li style="list-style-type:disc;">알고리즘 복잡도 분석(Algorithm complexity analysis)</li>
            <ul><ul>
                <li style="list-style-type:circle;">특정한 알고리즘의 문제의 크기(n)에 따른 실행시간(시간 복잡도)과 추가적으로 필요한 저장공간(공간 복잡도)을 분석하는 것을 알고리즘 복잡도 분석이라고 한다.</li>
                <li style="list-style-type:circle;">Big-O notation을 이용하여 나타낸다. (주로 고차항만으로 표시한다.)</li>
                <li style="list-style-type:circle;">최선/평균/최악(best/average/worst case)의 3가지 경우를 모두 분석한다.</li>
            </ul></ul>
            <li style="list-style-type:disc;">시간 복잡도 분석</li>
            <ul><ul>
                <li style="list-style-type:circle;">비교, 대입, 기본연산이 단위시간(1)의 실행시간을 가진다고 가정한다.</li>
                <li style="list-style-type:circle;">자료구조(리스트, 튜플, 문자열 등)에 대한 연산은 별도로 어떠한 시간 복잡도를 가지는지 분석해야 한다.</li>
            </ul></ul>
            <li style="list-style-type:disc;">공간 복잡도</li>
            <ul><ul>
                <li style="list-style-type:circle;">기본 자료형(int, float, 글자)이 단위 저장 공간을 가진다고 가정한다.</li>
            </ul></ul>
        </ul></ul>
        <br>
        <hr>
        <h3>2. Big-O notation의 정의</h3>
        <ul><ul>
            <li style="list-style-type:disc;">모든 n≥k에 대하여, 어떠한 상수 c와 k가 0≤f(n)≤cg(n)을 만족할 때, f(n)=O(g(n))으로 표시한다. 이 때, c와 k는 n과 독립적으로 고정된 상수여야 한다.</li>

            <h4>D. 선택 정렬의 복잡도 분석</h4>
            <ul><ul>
                <li style="list-style-type:disc;">선택 정렬은 리스트에 n개의 아이템이 있을 경우, 다음 단계를 n-1회 반복하는 알고리즘이다.</li>
                <ul><ul>
                    <li style="list-style-type:circle;">정렬되지 않은 부분에서 가장 작은 수를 찾음.</li>
                    <li style="list-style-type:circle;">가장 작은 수와 정렬되지 않은 부분의 첫번째 수를 교환.</li>
                </ul></ul>
                <li style="list-style-type:disc;">시간 복잡도</li>
                <ul><ul>
                    <li style="list-style-type:circle;">비교횟수: $$(n-1)+(n-2)+⋯+1=\frac{(n-1)n}{2}\,\,→\,\,O(n^2)$$</li>
                </ul></ul>
                <li style="list-style-type:disc;">공간 복잡도</li>
                <ul><ul>
                    <li style="list-style-type:circle;">교환을 위한 변수 1개 $$O(1)$$</li>
                </ul></ul>
            </ul></ul>
        </ul></ul>
        <br>

        <h3>4. 병합 정렬(merge sort)</h3>
        <ul><ul>
            <h4>A. 병합 정렬(merge sort)</h4>
            <ul><ul>
                <img src="https://blog.boot.dev/img/800/merge_sort_gif.gif" alt="" loading="lazy" align="left" style="width:500px;margin-right:100%;">
                <li style="list-style-type:disc;">분할정복법(divide&conquer)에 의한 정렬 알고리즘이다. 다시말해, 정렬해야 하는 데이터들을 받아 분할하고 분할한 것에 대해 각각 정렬한 뒤 합쳐서(merge) 전체적으로 정렬된 데이터들을 리턴한다.</li>
                <li style="list-style-type:disc;">n개의 아이템이 있는 리스트를 정렬하기 위하여,</li>
                <ul><ul>
                    <li style="list-style-type:circle;">n=1: base case(리스트는 이미 정렬되어 있음)</li>
                    <li style="list-style-type:circle;">n>1: merge_sort()와 merge() 함수를 이용해 정렬</li>
                    <ul><ul>
                        <text>▶ 왼쪽 절반을 merge_sort()를 이용해 정렬<br></text>
                        <text>▶ 오른쪽 절반을 merge_sort()를 이용해 정렬<br></text>
                        <text>▶ 정렬된 데이터를 merge()를 이용해 병합</text>
                    </ul></ul>
                </ul></ul>
            </ul></ul>
            <h4>B. 병합(merging): merge() 함수에서 일어나는 일</h4>
            <ul><ul>
                <li style="list-style-type:disc;">병합 정렬(merge sort) 알고리즘의 과정 중에서 이미 정렬된 두 개의 리스트를 합쳐 하나의 정렬된 리스트로 만드는 과정.</li>
                <li style="list-style-type:disc;">두 리스트가 이미 각각 정렬된 상태이기 때문에 각 리스트에서 가장 작은값 찾기를 시전할 필요 없이 그냥 각 리스트 맨 앞의 데이터를 가져오면 각 리스트에서 가장 작은값이 된다. 이렇게 가져온 두가지 데이터를 비교해서 더 작은 값을 새 리스트에 먼저 넣고, 다시 각 리스트에서 맨 앞 데이터 가져오기 및 비교 후 추가를 반복하면 merging함과 동시에 sorting이 된다.</li>
                <ul><ul>

        <h3>5. 퀵 정렬(quick sort)</h3>
        <ul><ul>
            <h4>A. 퀵 정렬(quick sort)</h4>
            <ul><ul>
                <img src="./Imgs for W15Note/quick sort 1.png" alt="" loading="lazy" align="left" style="width:500px;margin-right:100%;">
                <img src="https://upload.wikimedia.org/wikipedia/commons/f/fe/Quicksort.gif?20140912163236" alt="" loading="lazy" align="left" style="width:500px;margin-right:100%;">
                <ul><ul>
                    <li style="list-style-type:disc;">평균적으로 가장 빠른(시간복잡도는 안좋을지 몰라도 실제 실행시간이 가장 빠른) 정렬 알고리즘으로 가장 빈법하게 사용된다.</li>
                    <ul><ul>
                        <li style="list-style-type:circle;">merge sort, heap sort에 비해 평균 2-3배 정도 빠르다.</li>
                    </ul></ul>
                    <li style="list-style-type:disc;">분할정복법을 이용한다. (일반적으로 재귀를 사용하여 구현된다.)</li>
                    <ul><ul>
                        <li style="list-style-type:circle;">순서1: partition 알고리즘: pivot value보다 큰 수와 작은 수의 영역으로 분할하고, piviot보다 큰 수는 큰 수의 영역으로, 작은 수는 작은 수의 영역으로 보낸다.</li>
                        <ul><ul>
                            <text>▶ pivot value는 일반적으로 가장 앞, 가운데, 끝의 값 중 하나를 사용한다.</text>
                        </ul></ul>
                        <li style="list-style-type:circle;">순서2: 순서1이 완료되면 순서1에서 pivot이었던 데이터는 정렬 후에도 제자리를 유지하게 된다. 이제 두개로 분할된 영역에서 각각 또 pivot을 뽑아서 재귀적으로 정렬한다.</li>
                    </ul></ul>
                </ul></ul>
            </ul></ul>

        <h3>6. 정렬 알고리즘 비교</h3>
        <ul><ul>
            <img src="./Imgs for W15Note/Comparison sorting algorithms.png" alt="" loading="lazy" align="left" style="width:500px;margin-right:100%;">
            <ul><ul>
                <li style="list-style-type:disc;">Quick sort는 pivot이 중앙값으로 잡힐때(Best, Average), Merge sort처럼 작용해서 O(nlogn) time이 걸리고, pivot이 양 side 끝값으로 잡힐때(Worst), Selection sort처럼 작용해서 O(n^2) time이 걸린다.</li>
            </ul></ul>
        </ul></ul>
