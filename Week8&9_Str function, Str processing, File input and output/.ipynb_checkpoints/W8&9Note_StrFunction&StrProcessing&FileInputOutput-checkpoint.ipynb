                <h4>A. 문자열 함수(일부)</h4>
                <ul><ul>
                    <li style="list-style-type:disc;">문자열 함수의 용법: str.function(arguments)와 같은 형식으로 쓴다.</li>
                    <ul><ul>
                        <li style="list-style-type:circle;">나누기/붙이기: split( ), splitlines( ), join( )</li>
                        <li style="list-style-type:circle;">문자열 검색/치환: find( ), rfind( ), startswith( ), endswith( ), count( ), replace( ), index( ), rindex( )</li>
                        <li style="list-style-type:circle;">서식화: rjust( ), ljust( ), center( ), format( )</li>
                        <li style="list-style-type:circle;">대소문자: title( ), capitalize( ), lower( ), upper( )</li>
                        <li style="list-style-type:circle;">문자 확인: isalnum( ), isdecimal( ), isalpha( )</li>
                        <li style="list-style-type:circle;">공백문자 제거: strip( ), rstrip( ), lstrip( )</li>
                    </ul></ul>
                    <li style="list-style-type:disc;">str에 위치할 수 있는 것들</li>
                    <ul><ul>
                        <li style="list-style-type:circle;">문자열</li>
                        <li style="list-style-type:circle;">문자열을 가르키는 변수</li>
                        <li style="list-style-type:circle;">문자열을 반환하는 함수</li>
                    </ul></ul>
                    <li style="list-style-type:disc;">반환값: 대다수의 함수들은 문자열을 반환</li>
                    <ul><ul>
                        <li style="list-style-type:circle;">예외: split( ), splitlines( )는 리스트를 반환한다.</li>
                    </ul></ul>
                    <li style="list-style-type:disc;"><a href="https://docs.python.org/3.7/library/stdtypes.html#string-methods">추가적인 내용</a></li>
                </ul></ul>
                <h4>B. split( ) : 구분자를 기준으로 분리된 문자열의 리스트를 반환</h4>
                <ul><ul>
                    <li style="list-style-type:disc;">str.split(sep=None, maxsplit=-1)</li>
                    <ul><ul>
                        <li style="list-style-type:circle;">문자열이 들어왔을 때, 단어로 쪼개서 리스트로 반환하는 함수이다. 문자열을 쪼갤 때는 함수에서 sep 변수로 받은 값에 해당하는 문자를 기준으로 쪼개게 된다. 만약 sep이 주어지지 않으면, widespace(문자열의 공백 및 , 를 포함한 파이썬에서 정해둔 다양한 문자열의 요소)를 기준으로 쪼개게 된다.</li>
                        <li style="list-style-type:circle;">maxsplit이 주어지면, 쪼개는 횟수를 maxsplit의 값만큼 진행한다. 예를 들어 maxsplit=1이면, 한 문장을 한번만 쪼갠다. 그 결과 두개의 문자열이 리턴된다. (즉, split으로 리턴되는 리스트의 원소의 개수는 maxsplit+1개이다.)</li>
                        <li style="list-style-type:circle;">maxsplit이 주어지지 않거나 -1이면, split의 횟수에 제한을 주지 않는 것이다. 그 결과 문자열을 처음부터 끝까지 다 돌면서 해당하는 기준문자에 대해 문자열을 쪼갠다.</li>
                        <li style="list-style-type:circle;">sep이 주어지면, 그 sep을 기준으로 문자열을 쪼갠다. 예를 들어 ‘1,,2’.split(‘,’)이라 쓰면, [‘1’, ‘’, ‘2’]가 리턴된다.</li>
                        <li style="list-style-type:circle;">sep으로 다중 문자열을 주는 것도 가능하다. 예를 들어 ‘1<>2<>3’.split(‘<>’)이라 쓰면, [‘1’, ‘2’, ‘3’]가 리턴된다.</li>
                        <li style="list-style-type:circle;">str이 빈 문자열 ‘’인 경우 split함수는 [‘’]을 리턴한다.</li>
                    </ul></ul>
                </ul></ul>
                <h4>C. splitlines( ) : 개행문자로 분리된 문자열의 리스트를 반환</h4>
                <ul><ul>
                    <li style="list-style-type:disc;">str.splitlines([keepends])</li>
                    <ul><ul>
                        <li style="list-style-type:circle;">이 함수는 line boundaries(\n: 줄 바꾸기)에서 문자열을 쪼개서 line들을 만들고, 이 line들을 리스트에 넣어 반환하는 함수이다.</li>
                        <li style="list-style-type:circle;">함수의 parameter인 keepends는 Boolean값으로, True 또는 False를 입력하도록 한다. 그런데 이 parameter는 입력해도 되고, 입력하지 않아도 된다. 그 표시로 [keepends]라고 쓴 것이다. keepends를 입력하지 않거나 False를 입력하면, 리스트 안의 원소들(line들)은 \n을 기준으로 쪼개진게 들어가되 \n은 없어지고, keepends를 True로 입력하면, 리스트 안의 원소들(line들)은 \n을 기준으로 쪼개진게 들어가되 각 lines의 끝에 \n이 온전히 보존된 채로 들어가게 된다.</li>
                        <li style="list-style-type:circle;">참고: split(‘\n’)과 유사하나, 마지막 빈 줄을 남기지 않는다는 점에서 차이가 있다.</li>
                    </ul></ul>
                </ul></ul>

                <h4>F. strip( ) : 앞뒤로 지정된 문자열을 제거</h4>
                <ul><ul>
                    <li style="list-style-type:disc;">str.strip([chars])</li>
                    <ul><ul>
                        <li style="list-style-type:circle;">input parameter인 [chars]는 생략가능하다.</li>
                        <li style="list-style-type:circle;">[chars]를 주면, str을 복사하고 문장 맨 앞에서부터 chars에 해당하지 않는 문자가 나올때까지 문자들을 제거하여 나간다. 또한, 맨 뒤에서부터 chars에 해당하지 않는 문자가 나올 때까지 chars를 제거하여 리턴한다. [chars]를 주지 않거나 None을 주면, str을 복사하고, str의 맨 앞에서부터 공백(빈칸, 띄어쓰기, tab, 줄바꿈)이 없을 때까지 공백을 제거하여나가고, 맨 뒤에서부터 공백이 없을 때까지 공백을 제거하여 리턴한다.</li>
                        <li style="list-style-type:circle;">참고: 파일/네트워크에서 한 줄을 읽은 후, 공백을 제거하기 위해 주로 사용한다.</li>
                    </ul></ul>
                </ul></ul>
                <h4>G. rstrip( ), lstrip( ) : 앞 혹은 뒤의 지정된 문자열 제거</h4>
                <ul><ul>
                    <li style="list-style-type:disc;">str.rstrip([chars])</li>
                    <ul><ul>
                        <li style="list-style-type:circle;">[chars]를 주면, str을 복사하고 문장 맨 뒤에서부터 chars에 해당하지 않는 문자가 나올 때까지 chars를 제거하여 리턴한다. [chars]를 주지 않거나 None을 주면, str을 복사하고, str의 맨 뒤에서부터 공백(빈칸, 띄어쓰기, tab, 줄바꿈)이 없을 때까지 공백을 제거하여 리턴한다.</li>
                    </ul></ul>
                    <li style="list-style-type:disc;">str.lstrip([chars])</li>
                    <ul><ul>
                        <li style="list-style-type:circle;">[chars]를 주면, str을 복사하고 문장 맨 앞에서부터 chars에 해당하지 않는 문자가 나올때까지 문자들을 제거하여 리턴한다. [chars]를 주지 않거나 None을 주면, str을 복사하고, str의 맨 앞에서부터 공백(빈칸, 띄어쓰기, tab, 줄바꿈)이 없을 때까지 공백을 제거하여 리턴한다.</li>
                    </ul></ul>

            <h3>2. 문자열 서식화</h3>
            <ul><ul>
                <h4>A. python에서의 문자열 서식화</h4>
                <ul><ul>
                    <li style="list-style-type:disc;">서식문자열에 따라, (변수에 저장된) 값의 자리수, 표현방식을 서식화하는 것을 말한다.</li>
                    <li style="list-style-type:disc;">서식화된 ‘문자열’을 반환한다는 것은 변수에 대입하거나, printf( ) 함수를 이용하여 출력한다는 것을 의미한다.</li>
                    <ul><ul>
                        <li style="list-style-type:circle;">참고: C언어의 printf( ) 함수가 아닌, sprintf( ) 함수에 해당한다.</li>
                    </ul></ul>
                </ul></ul>
                <h4>B. python에서 지원하는 문자열 서식화 방식</h4>
                <ul><ul>
                    <li style="list-style-type:disc;">Printf-style string formatting (all python versions)</li>
                    <ul><ul>
                        <li style="list-style-type:circle;">C언어의 printf( ) 함수와 유사한 서식문자열을 지원한다.</li>
                    </ul></ul>
                    <li style="list-style-type:disc;">format( ) function with format string syntax (python3, python 2.6+)</li>
                    <ul><ul>
                        <li style="list-style-type:circle;">서식문자열의 표현이 더 풍부해졌다.</li>
                    </ul></ul>
                    <li style="list-style-type:disc;">Literal string interpolation (python 3.6+)</li>
                    <ul><ul>
                        <li style="list-style-type:circle;">서식문자열에 수식 혹은 변수의 이름을 적을 수 있는 기능을 추가하였다.</li>
                    </ul></ul>
                    <li style="list-style-type:disc;">Template string (python 2+): printf-style과 유사하다.</li>
                </ul></ul>
                <h4>C. 문자열 서식화 방법 1: Printf-style string formatting (%operator) 소개</h4>
                <ul><ul>
                    <li style="list-style-type:disc;">사용형태: format % values</li>
                    <ul><ul>
                        <li style="list-style-type:circle;">format: 서식 문자열로, 다음 형태의 변환명시자(conversion specifier)를 포함한다.</li>
                        <ul><ul>
                            <p>▶ %로 시작하고, 변환형을 나타내는 하나의 문자로 끝나는 변환명시자.</p>
                            <p>▶ mapping key, conversion flag, 최소 문자 폭, 정확도를 명시하는 변환명시자.</p>
                        </ul></ul>
                        <li style="list-style-type:circle;">values: 다음 중 하나의 형식을 사용한다.</li>
                        <ul><ul>
                            <p>▶ 한 개의 값(서식 문자열이 하나의 값을 요구할 때).</p>
                            <p>▶ 서식문자열이 요구하는 아이템 수와 같은 수의 값이 있는 튜플.</p>
                            <p>▶ 한 개의 사전(dict)과 같은 매핑 객체(mapping object)</p>
                        </ul></ul>
                    </ul></ul>
                    <li style="list-style-type:disc;">기능: 서식문자열에 따라 values에 주어진 값들을 원하는 형식의 ‘문자열’로 변환한다.</li>
                    <li style="list-style-type:disc;">단점: 튜플, 사전, 객체 등 복잡한 자료형의 값을 출력할 때 제한점이 많다.</li>
                    <li style="list-style-type:disc;">%operator는 string formatting operator 또는 interpolation operator라고도 불린다.</li>
                    <li style="list-style-type:disc;">참고: C언어의 printf( ) 혹은 sprintf( ) 함수에서 서식문자열과 유사하다.</li>
                </ul></ul>
                <h4>D. 문자열 서식화 방법 1: 변환 명시자(Conversion specifier) 형식</h4>
                <ul><ul>
                    <li style="list-style-type:disc;">%로 시작해서, 변환형을 나타내는 하나의 문자로 끝나야 한다. 즉, %(optional)(conversionoperator)의 형식으로 표현되어야 한다.</li>
                    <li style="list-style-type:disc;">(Optional)자리에는 mapping key를 쓸 수 있다.</li>
                    <ul><ul>
                        <li style="list-style-type:circle;">( )안에 문자들로 주어진다.</li>
                        <li style="list-style-type:circle;">맵핑 객체 형태의 values에서 키가 ( )안의 문자들인 값으로 대체된다.</li>
                    </ul></ul>
                    <li style="list-style-type:disc;">(Optional)자리에 conversion flag (일부)를 쓸 수도 있다.</li>
                    <ul><ul>
                        <li style="list-style-type:circle;">‘0’: 수의 경우, 남는 공간이 0으로 채워진다.</li>
                        <li style="list-style-type:circle;">‘-‘: 변환된 문자열이 왼쪽에 정렬된다.</li>
                        <li style="list-style-type:circle;">conversion flag를 지정하지 않으면 문자열이 우측에 정렬된다.</li>
                    </ul></ul>
                    <li style="list-style-type:disc;">(Optional)자리에 최소 문자 폭: 출력되는 값의 최소 문자의 수를 지정하는 정수를 쓸 수도 있다.</li>
                    <li style="list-style-type:disc;">(Optional)자리에 정확도: ‘.’와 숫자로, 소수 부분의 자릿수를 명시할 수도 있다.</li>
                    <li style="list-style-type:disc;">(Optional)자리에 length modifier를 쓰는 언어도 있으나, python에서는 사용되지 않는다.</li>
                    <table align="left" style="width:500px;margin-right:100%;">
                        <thead>
                            <tr><th>Conversion</th><th>Meaning</th></tr>
                        </thead>
                        <tbody>
                            <tr><td>‘d’ or ‘i’</td><td>value를 문자열 안에 정수로 변환해서 집어넣을 때 사용.</td></tr>
                            <tr><td>‘o’</td><td>value를 문자열 안에 8진수로 변환해서 집어넣을 때 사용.</td></tr>
                            <tr><td>‘u’</td><td>Obsolete type – it is identical to ‘d’.</td></tr>
                            <tr><td>‘x’ or ‘X’</td><td>value를 문자열 안에 16진수로 변환해서 집어넣을 때 사용한다. (lowercase(‘x’) or uppercase(‘X’)</td></tr>
                            <tr><td>‘e’ or ‘E’</td><td>value를 문자열 안에 실수로 변환해서 집어넣되, 매우 작거나 매우 커서 지수승으로 표기해야 할 때 사용. (lowcase(‘e’) or uppercase(‘E’).)</td></tr>
                            <tr><td>‘f’ or ‘F’</td><td>value를 문자열 안에 실수로 변환해서 집어넣을 때 사용.</td></tr>
                            <tr><td>‘g’ or ‘G’</td><td>value에 정수가 들어오면 정수로, 매우 커거나 작아서 지수승으로 표기해야 하면 지수승으로, 실수로 들어오면 실수로 변환해서 문자열 안에 집어넣어준다.</td></tr>
                            <tr><td>‘c’</td><td>value를 한 개의 문자로 변환하여 문자열 안에 집어넣을 때 사용. (숫자 한 개나 문자 한 개에 대해 사용한다.)</td></tr>
                            <tr><td>‘r’, ‘s’, ‘a’</td><td>value를 문자열로 변환하여 문자열 안에 집어넣을때 사용한다. 이렇게 쓰는 것은 각각 repr( ), str( ), ascii( )와 같은 기능을 한다.</td></tr>
                            <tr><td>‘%’</td><td>문자열 안에 %를 쓰고 싶으면 %%와 같이 사용하면 된다.</td></tr>
                        </tbody>
                    </table>
                </ul></ul>

                    <li style="list-style-type:disc;">사용형태: string.format(arguments)</li>

                        <li style="list-style-type:circle;">string: 서식 문자열</li>

                    </ul></ul>
                    <li style="list-style-type:disc;">참고: 서식화를 지정하는 서식 명시자(formatting specifier)는 printf 형태의 변환 명시자와 유사하지만, 그 기능이 확장되었고 사용이 조금 더 편리하도록 변경되었다.</li>

        <h3>4.	텍스트 파일 쓰기</h3>
            <ul><ul>
                <li style="list-style-type:disc;">코드 설명</li>
                <ul><ul>
                    <li style="list-style-type:circle;">아래의 코드는 코드파일과 같은 주소에 있는 example.txt라는 이름의 파일에 문자열 something을 쓰는 알고리즘을 수행한다. 코드파일과 같은 주소에 example.txt가 없으면, example.txt 파일을 자동으로 만들어서 알고리즘을 수행한다.</li>
                    <ul><ul>
                        <p>▶ 일반적인 방법</p>
                        <ul><ul>
                            <table align="left" style="width:250px;margin-right:100%;">
                                <tbody>
                                    <tr><td style="text-align:left;">fileobj = open(‘example.txt’, ‘wt’)<br>print(‘something’, file=fileobj)<br>fileobj.close( )</td></tr>
                                </tbody>
                            </table>
                        </ul></ul>
                        <p>▶ with를 사용하는 방법 (close( )를 호출할 필요가 없다.)</p>
                        <ul><ul>
                            <table align="left" style="width:250px;margin-right:100%;">
                                <tbody>
                                    <tr><td style="text-align:left;">with open(‘example.txt’, ‘wt’) as fileobj:<br>print(‘something’, file=fileobj)</td></tr>
                                </tbody>
                            </table>
                        </ul></ul>
                    </ul></ul>
                </ul></ul>
